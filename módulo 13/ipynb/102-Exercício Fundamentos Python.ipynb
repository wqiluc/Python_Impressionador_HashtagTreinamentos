### 1- Imposto a pagar no Lucro Presumido

- 5% sobre faturamento de ISS (mensal)
- 0,65% de PIS sobre faturamento, (mensal)
- 3% de COFINS sobre faturmaneto, (mensal)
- 4.8% de IR (trimestral)
- 10% de IR Adicional sobre o que ultrapassar 20mil do faturamento (trimestral)
- CSLL: 2,88% sobre faturamento (trimestral)

```python
# Dicionário original com os valores de faturamento
faturamento = {
    'jan': 'R$ 95.141,98',
    'fev': 'R$ 95.425,16',
    'mar': 'R$ 89.716,31',
    'abr': 'R$ 78.459,99',
    'mai': 'R$ 71.087,28',
    'jun': 'R$ 83.911,06',
    'jul': 'R$ 56.467,26',
    'ago': 'R$ 88.513,58',
    'set': 'R$ 66.552,49',
    'out': 'R$ 80.164,07',
    'nov': 'R$ 66.964,33',
    'dez': 'R$ 71.525,25',
}

def processar_faturamento(dados_faturamento):
    """
    Processa os dados de faturamento para calcular os impostos mensais e trimestrais.
    
    :param dados_faturamento: Dicionário com os valores de faturamento em formato string.
    :return: Dicionário contendo os resultados limpos e calculados.
    """
    
    # você precisa inserir no sistema um dicionário no formato:
    # resultado = {
    #     'mes': (faturamento, imposto_mensal, imposto_trimestral),
    # }

    dicionario_resultado = { }

    for (mes, valor_do_mes) in (dados_faturamento.items()):
        valor_limpo = valor_do_mes.replace('R$ ', '').replace('.', '').replace(',', '.')
        valor_limpo = float(valor_limpo)
        
        # Cálculos de impostos
        iss = 0.05 * valor_limpo
        pis = 0.01 * valor_limpo
        cofins = 0.04 * valor_limpo
        
        imposto_mensal = valor_limpo * 0.05
        imposto_trimestral = valor_limpo * 0.15
        
        dicionario_resultado[mes] = (valor_limpo, imposto_mensal, imposto_trimestral)
        
    return (dicionario_resultado)


def exibir_relatorio(resultado):
    """
    Exibe o relatório formatado a partir do dicionário de resultados.
    
    :param resultado: Dicionário com os dados processados.
    """
    for (mes, dadosmes) in resultado.items():
        print(f'Mês: {mes.capitalize()}')
        print(f'Faturamento: R$ {dadosmes[0]:.2f}')
        print(f'Imposto Mensal: R$ {dadosmes[1]:.2f}')
        print(f'Imposto Trimestral: R$ {dadosmes[2]:.2f}')
        print('---')

if (__name__ == "__main__"):
    resultado_processado = processar_faturamento(faturamento)
    exibir_relatorio(resultado_processado)
```

### 3 - Extração dos links de Download de Vídeos do Vimeo

- Precisamos pegar os links em 1080p, 720p e 540p para importar os vídeos para uma nova plataforma

```python
# para pegar o dicionario do vimeo, use:
from database import dicionario_vimeo

# você precisa chegar em:
# videos = [
#     {'uri': video_uri, 'nome': nome_video, 'duracao': duracao, 'link540p': link540p, 'link720p': link720p, 'link1080p': link1080p},
# ]

lista_videos_vimeo = list()

for dadosvideos in dicionario_vimeo['data']:
    video_uri = dadosvideos['uri']
    nome_video = dadosvideos['name']
    duracao = dadosvideos['duration']
    link540p = dadosvideos['files'][0]['link']
    link720p = dadosvideos['files'][1]['link']
    link1080p = dadosvideos['files'][2]['link']
    lista_videos_vimeo.append({'uri': video_uri, 'nome': nome_video, 'duracao': duracao, 'link em 540p': link540p, 'link em 720p': link720p, 'link em 1080p': link1080p})

for indicevideos, videos in (enumerate(lista_videos_vimeo)):
    print(f"URI do {indicevideos+1}º Vídeo: {videos['uri']}")
    print(f"Nome do {indicevideos+1}º Vídeo: {videos['nome']}")
    print(f"Duração do {indicevideos+1}º Vídeo: {videos['duracao']} segundos")
    print(f"Link em 540p: {videos['link em 540p']}")
    print(f"Link em 720p: {videos['link em 720p']}")
    print(f"Link em 1080p: {videos['link em 1080p']}")
    print('\n')
```
