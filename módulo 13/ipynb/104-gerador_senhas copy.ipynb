## 📋 Descrição
 
Este projeto implementa um **gerador de senhas aleatórias** em Python. O programa solicita ao usuário um comprimento desejado e cria uma senha robusta contendo obrigatoriamente:
 
- 🔡 Pelo menos **uma letra** (maiúscula ou minúscula)
- 🔢 Pelo menos **um número**
- 🔣 Pelo menos **um caractere especial**
- 🚫 **Sem espaços em branco**
Cada execução gera uma senha diferente, garantindo aleatoriedade total.

## 🧠 Como Funciona
 
O coração do projeto é a função `gerar_senha(comprimento)`:
 
```python
import random
import string
 
def gerar_senha(comprimento):
    """
    Gera uma senha aleatória com o comprimento especificado.
    Retorna None se o comprimento for inválido (< 4).
    """
    if comprimento < 4:
        return None
 
    letras = string.ascii_letters       # a-z + A-Z
    numeros = string.digits             # 0-9
    especiais = string.punctuation      # !@#$%^&*...
 
    # Garante pelo menos um de cada tipo
    senha = [
        random.choice(letras),
        random.choice(numeros),
        random.choice(especiais),
    ]
 
    # Preenche o restante com caracteres aleatórios
    todos = letras + numeros + especiais
    senha += [random.choice(todos) for _ in range(comprimento - 3)]
 
    # Embaralha para não ter padrão fixo no início
    random.shuffle(senha)
 
    return ''.join(senha)
```
 
### 🔍 Lógica passo a passo
 
| Etapa | Descrição |
|-------|-----------|
| 1️⃣ Validação | Verifica se o comprimento é ≥ 4 |
| 2️⃣ Garantia | Adiciona 1 letra, 1 número e 1 especial obrigatoriamente |
| 3️⃣ Preenchimento | Completa os caracteres restantes aleatoriamente |
| 4️⃣ Embaralhamento | Usa `random.shuffle()` para misturar tudo |
| 5️⃣ Retorno | Junta a lista em uma string e retorna |
 
---
 
## 📦 Bibliotecas Utilizadas
 
| Biblioteca | Uso |
|------------|-----|
| `random` | `choice()` para sortear caracteres, `shuffle()` para embaralhar |
| `string` | Conjuntos prontos: `ascii_letters`, `digits`, `punctuation` |
 
---
 
## 💡 Dicas de Implementação
 
> 🔹 Use `random.choice(sequência)` para selecionar um item aleatório de uma string ou lista.
 
> 🔹 Use `random.shuffle(lista)` para embaralhar os elementos de uma lista **in-place**.
 
> 🔹 Use `string.ascii_letters`, `string.digits` e `string.punctuation` para obter os conjuntos de caracteres sem precisar digitá-los manualmente.
 

## 📌 Requisitos
 
- Python 3.x
- Sem dependências externas

````python
import random
import string

def gerar_senha_aleatoria(tamanho=16):
    caracteres = string.ascii_letters + string.digits + string.punctuation + string.hexdigits
    senha_usuario = "".join(random.choice(caracteres) for _ in range(tamanho))
    return senha_usuario

def exibir_senha(tamanho=16):
    senha = gerar_senha_aleatoria(tamanho)
    print(f"Senha gerada: {senha}")

if (__name__ == "__main__"):
    exibir_senha(tamanho=16)
````
