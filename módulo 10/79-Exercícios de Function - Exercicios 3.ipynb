## 1. Cálculo do Percentual e da Lista de Vendedores

- Queremos criar uma function que consiga identificar os vendedores que bateram uma meta, mas além disso, consigo já me dar como resposta o cálculo do % da lista de vendedores que bateu a meta (para eu não precisar calcular manualmente depois)
- Essa function deve receber 2 informações como parâmetro: a meta e um dicionário com os vendedores e suas vendas. E me dar 2 respostas: uma lista com o nome dos vendedores que bateram a meta e o % de vendedores que bateu a meta.

```python
from cores import(Negrito,Reset,Verde)
meta = 10000
vendas = {
    'João': 15000,
    'Julia': 27000,
    'Marcus': 9900,
    'Maria': 3750,
    'Ana': 10300,
    'Alon': 7870,
    }
```

```python
def percentual_vendedores(meta=meta, vendedores_acima_meta=list()):
    vendedores_acima_meta = [ ]
    for indice_vendedor, dado_vendedor in enumerate(vendas.items()):
        if dado_vendedor[1] >= meta:
            vendedores_acima_meta.append([dado_vendedor[0], dado_vendedor[1]])
    percentual = (len(vendedores_acima_meta) / len(vendas)) * 100
    for indice_vendedor, vendedor in enumerate(vendedores_acima_meta[0:]):
        print(
            f"{Negrito}{indice_vendedor+1}º {Reset}"
            f"{Negrito}Vendedor:{Reset} {Verde}{vendedor[0]}{Reset} - "
            f"{Negrito}Valor Vendido:{Reset} {Verde}R${Reset}{Negrito}{vendedor[1]}{Reset} - "
            f"{Negrito}Percentual de Vendedores que bateram a meta:{Reset} {Verde}{percentual:.2f}%{Reset}")
```

```python
#aplique sua function nas informações para ver se está funcionando
resposta = percentual_vendedores(meta=10000, vendedores_acima_meta=list())
```
