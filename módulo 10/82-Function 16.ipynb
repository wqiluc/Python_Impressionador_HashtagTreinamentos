```python
def minha_soma(*numeros):
    print(numeros)
    soma = 0
    for indice_numero, numero in enumerate(numeros):
        soma += numero
        print(f"√çndice: {indice_numero+1}, N√∫mero: {numero}, Soma parcial: {soma}")
    return soma
```

```python
resultado_exemplo1 = minha_soma(1, 2, 3, 4, 5)
```

```python
def preco_final(preco, **adicionais):
    print(adicionais)
    if 'Desconto'.lower() in (adicionais):
        preco *= (1 - adicionais['desconto'])
    if 'garantia_extra'.lower().replace(' ', '') in (adicionais):
        preco += adicionais['garantia_extra'] 
    if 'imposto' in (adicionais):
        preco *= (1 + adicionais['imposto'])
    return preco
```

```python
resultado_exemplo2 = preco_final(1000, desconto=0.1, garantia_extra=150, imposto=0.2)
```
