#### 23. Faça um Programa para uma loja de tintas. O programa deverá pedir o tamanho em metros quadrados da área a ser pintada. Considere que a cobertura da tinta é de 1 litro para cada 6 metros quadrados e que a tinta é vendida em latas de 18 litros, que custam R\\$ 80,00 ou em galões de 3,6 litros, que custam R\\$ 25,00.

```python
from cores import (Negrito, Reset, Vermelho)

def loja_tintas():
    area_m2 = str(input(f"{Negrito}Qual a área em m² que você deseja pintar? {Reset}").strip().lower())
    try:
        area_m2 = float(area_m2)
        if (area_m2 <= 0):
            print(f"{Vermelho}Área inválida.❌ Por favor, insira um valor positivo.{Reset}")
            return loja_tintas()
        return area_m2
    except ValueError:
        print(f"{ValueError}")
        print(f"{Vermelho}Entrada inválida. Por favor, insira um número válido.{Reset}")
        return (loja_tintas())

def litros_necessarios(area_m2):
    litros = area_m2 / 6
    latas = litros / 18
    galoes = litros / 3.6
    return litros, latas, galoes

def custo_total(latas, galoes):
    custo_lata = 80.00
    custo_galao = 25.00
    total_latas = int(latas) * custo_lata
    total_galoes = int(galoes) * custo_galao
    return total_latas, total_galoes

def principal():
    area_m2 = loja_tintas()
    litros, latas, galoes = litros_necessarios(area_m2)
    total_latas, total_galoes = custo_total(latas, galoes)

    print(f"{Negrito}Para pintar {area_m2:.2f} m², você precisará de {litros:.2f} litros de tinta.{Reset}")
    print(f"{Negrito}Opção 1: Comprar {int(latas)+1} latas de 18 litros. Custo total: R$ {total_latas:.2f}{Reset}")
    print(f"{Negrito}Opção 2: Comprar {int(galoes)+1} galões de 3.6 litros. Custo total: R$ {total_galoes:.2f}{Reset}")

principal()
```
