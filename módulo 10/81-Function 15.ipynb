### Como "testar" erros e tratar exceções:

```python
from cores import(Negrito,Reset,Verde,Vermelho)
while True:
    def descobrir_servidor(email):
        try:
            posicao_a = email.index('@')
        except ValueError:
            return print(f"{Vermelho}E-mail inválido{Reset}")
        servidor = email[posicao_a:]
        if 'gmail' in (servidor):
            return f'{Verde}Seu domínio é: gmail✅{Reset}'
        elif 'hotmail' in (servidor) or 'outlook' in (servidor) or 'live' in (servidor):
            return f'{Verde}Seu domínio é: hotmail✅{Reset}'
        elif 'yahoo' in (servidor):
            return f'{Verde}Seu domínio é: yahoo✅{Reset}'
        elif 'uol' in (servidor) or 'bol' in (servidor):
            return f'{Verde}Seu domínio é: uol✅{Reset}'
        if '@' in email:
            return descobrir_servidor(email)
        else:
            return f'{Vermelho}Não determinado ❌{Reset}'
```

### Como "printar" um erro em uma function

```python
def descobrir_servidor(email):
    posicao_a = email.index('@')
    servidor = email[posicao_a:]
    if 'gmail' in (servidor):
        return f'{Verde}Seu domínio é: gmail✅{Reset}'
    elif 'hotmail' in (servidor) or 'outlook' in (servidor) or 'live' in (servidor):
        return f'{Verde}Seu domínio é: hotmail✅{Reset}'
    elif 'yahoo' in (servidor):
        return f'{Verde}Seu domínio é: yahoo✅{Reset}'
    elif 'uol' in (servidor) or 'bol' in (servidor):
        return f'{Verde}Seu domínio é: uol✅{Reset}'
    else:
        return f'{Vermelho}Domínio não determinado ❌{Reset}'
```

### Tratamento Completo:

```python
while True:
    try:
        email = str(input("Qual o seu e-mail? ")).strip().lower()
        print(descobrir_servidor(email))
        break
    except ValueError:
        print(f'{Vermelho}E-mail inválido ❌{Reset}')
```
