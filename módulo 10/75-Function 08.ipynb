- <strong>Nesse caso, voc√™ n√£o √© obrigado a passar o valor para usar a fun√ß√£o, pode usar o valor padr√£o.</strong>

- Como vimos, o <strong>sort()</strong> para listas tem um argumento padr√£o. O <strong>reverse = False √© padr√£o</strong>, ent√£o a ordem √© crescente. Caso o usu√°rio queira fazer em ordem decrescente, o reverse deve ser alterado para <strong>True‚úÖ</strong>

```python
from cores import(Negrito,Reset,Amarelo)

produtos = ['apple tv', 'mac', 'iphone x', 'iPad', 'apple watch', 'mac book', 'airpods']

produtos.sort()
# ou sorted(produtos)
print(produtos)

produtos.sort(reverse = True) #Par√¢metro reverse alterado p/ True (padr√£o=False)
print(produtos)
```
