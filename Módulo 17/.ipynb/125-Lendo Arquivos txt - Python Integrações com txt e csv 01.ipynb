## 📂 1. Abrindo Arquivos

O ponto de partida para trabalhar com qualquer arquivo `.txt` é abri-lo com `open()`, informando o que se pretende fazer com ele:

| Modo 🔑 | Abre para 🔓 |
|---|---|
| `'r'` | leitura (**r**ead) |
| `'w'` | escrita (**w**rite) — sobrescreve o conteúdo existente |
| `'a'` | adição (**a**ppend) — acrescenta ao final, sem apagar o que já existe |

## 📖 2. Lendo o Conteúdo de um Arquivo

Com o arquivo aberto, agora podemos efetivamente ler o conteúdo dele:

| Método 🔑 | O que retorna 🔓 |
|---|---|
| `.read()` | o conteúdo inteiro do arquivo como uma única string |
| `.readlines()` | uma lista, com cada linha do arquivo como um item |

## 🔒 4. Fechando o Arquivo

Quando você abre um arquivo, ele permanece aberto no Python até você fechar. Existem 2 formas de fazer isso:

| Forma 🔑 | Como funciona 🔓 |
|---|---|
| `.close()` | fecha o arquivo manualmente — precisa ser chamado explicitamente, depois de usar o arquivo |
| `with open(...) as arquivo:` | fecha o arquivo sozinho ao final do bloco, mesmo que ocorra um erro no meio do caminho |

> 💡 **Dica:** prefira sempre a estrutura `with` — ela fecha o arquivo automaticamente mesmo que aconteça um erro durante a leitura ou escrita, o que `.close()` sozinho não garante.

## 🎯 5. Desafio: Funil de Vendas da Hashtag

Na Hashtag, sempre analisamos o nosso "Funil de Vendas". Para isso, rastreamos de onde veio os alunos por meio de um código:

| Código 🔑 | Origem 🔓 |
|---|---|
| `hashtag_site_org` | site da Hashtag |
| `hashtag_yt_org` | YouTube da Hashtag |
| `hashtag_ig_org` | Instagram da Hashtag |
| `hashtag_igfb_org` | Instagram ou Facebook da Hashtag |

> ⚠️ **Atenção:** códigos diferentes desses são códigos de **anúncio** — orgânico é tudo aquilo que não veio de anúncios.

**Objetivos:**
- Quantos alunos vieram de anúncio e quantos vieram "orgânico"
- Qual a melhor fonte orgânica de alunos

No nosso sistema, conseguimos exportar um txt com as informações dos alunos, conforme o arquivo `Alunos.txt`.<br>
(Os dados foram gerados aleatoriamente para simular uma situação real, já que não podemos fornecer os dados reais dos alunos por questões de segurança)

> ✅ No final, para treinar, vamos escrever todas essas respostas em um novo arquivo txt.

```python
from cores import *

with open("../.spec/Lendo Arquivos txt - Alunos.txt", "r", encoding="latin-1") as arquivo:
    linhas = arquivo.readlines()

# as 4 (0-3) primeiras linhas são cabeçalho 
# (título, total de clientes, linha em branco, nome das colunas)
linhas_alunos = linhas[4:]

# só esses 4 prefixos contam como orgânico (conforme o enunciado do desafio) -> qualquer outro código é anúncio
fontes_organicas = {
    "hashtag_site_org": f"Site",
    "hashtag_yt_org": f"YouTube",
    "hashtag_igfb_org": f"Instagram ou Facebook",
    "hashtag_ig_org": f"SOMENTE Instagram",
}

qtd_organico = 0
qtd_anuncio = 0
contagem_fontes = {fonte: 0 for indice_fonte, fonte in enumerate(fontes_organicas.values())}

for indice_linha, linha in enumerate(linhas_alunos):
    linha = linha.strip().strip('"')
    email, origem = linha.split(",", 1)
    origem = origem.replace("origemurl:", "")
    codigos = origem.split(";")

    # primeiro código orgânico encontrado entre os touchpoints do aluno (first-touch)
    fonte_encontrada = None #ainda
    for indice_codigo, codigo in enumerate(codigos):
        for prefixo, nome_fonte in (fontes_organicas.items()):
            if codigo.startswith(prefixo):
                fonte_encontrada = nome_fonte
                break
            else:
                if (fonte_encontrada):
                    break

    if fonte_encontrada:
        qtd_organico += 1
        contagem_fontes[fonte_encontrada] += 1
    else:
        qtd_anuncio += 1

melhor_fonte = max(contagem_fontes, key=contagem_fontes.get)

print(f"{CinzaClaro}Total de alunos:{Reset} {VerdeClaro}{len(linhas_alunos)}{Reset}")
print(f"{CinzaClaro}Vieram de anúncio:{Reset} {VermelhoClaro}{qtd_anuncio}{Reset}")
print(f"{CinzaClaro}Vieram orgânico:{Reset} {VerdeClaro}{qtd_organico}{Reset}")
print(f"{CinzaClaro}Melhor fonte orgânica:{Reset} {MagentaClaro}{melhor_fonte} ({contagem_fontes[melhor_fonte]} alunos){Reset}")
for indice_fonte, fonte in enumerate(fontes_organicas.values()):
    qtd = contagem_fontes[fonte]
    print(f"{CinzaClaro}{fonte}:{Reset} {qtd}")

# escrever as respostas em um novo arquivo txt para enviar pra a subpasta .spec
with open("../.spec/Resultado Funil de Vendas.txt", "w", encoding="utf-8") as arquivo:
    arquivo.write(f"Total de alunos: {len(linhas_alunos)}\n")
    arquivo.write(f"Vieram de anúncio: {qtd_anuncio}\n")
    arquivo.write(f"Vieram orgânico: {qtd_organico}\n")
    arquivo.write(f"Melhor fonte orgânica: {melhor_fonte} ({contagem_fontes[melhor_fonte]} alunos)\n")
    for indice_fonte, fonte in enumerate(fontes_organicas.values()):
        qtd = contagem_fontes[fonte]
        arquivo.write(f"  {fonte}: {qtd}\n")
```
