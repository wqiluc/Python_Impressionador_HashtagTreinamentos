## 🧾 1. O que é uma Nota Fiscal em XML

Existem dois tipos de nota fiscal eletrônica bem diferentes por trás do XML:

| Tipo 🔑 | Nome técnico 🔓 | Representação em PDF |
|---|---|---|
| Nota de **produto** | NF-e (Nota Fiscal eletrônica) | DANFE |
| Nota de **serviço** | NFS-e (Nota Fiscal de Serviço eletrônica) | Nota do município (ex.: Nota Carioca) |

> ⚠️ **Atenção:** apesar de as duas serem "nota fiscal", a estrutura interna do XML é completamente diferente entre elas — por isso vamos precisar de uma função pra ler cada tipo.

Na subpasta `.spec` deste módulo tem um exemplo de cada:

| Arquivo 🔑 | Tipo 🔓 |
|---|---|
| `DANFEBrota.xml` / `DANFEBrota.pdf` | NF-e de produto (plantas) |
| `DANFENespresso.xml` / `DANFENespresso.pdf` | NF-e de produto (cápsulas de café) |
| `NotaCariocaHashtag.xml` / `NotaCariocaHashtag.pdf` | NFS-e de serviço |

## 📥 2. Lendo um XML com xmltodict

| Função/Método 🔑 | O que faz 🔓 |
|---|---|
| `xmltodict.parse(arquivo)` | converte o conteúdo do XML em um dicionário Python (aninhado, um dentro do outro) |
| `open(nota, 'rb')` | abre o arquivo em modo binário — evita problema de encoding, já que o próprio XML declara o encoding dele no cabeçalho |

> 💡 **Dica:** dá pra fazer um `print(documento)` logo depois do `parse()` só pra explorar a estrutura antes de sair navegando nas chaves do dicionário.

## 🛒 3. Função para Ler uma NF-e de Produto (DANFE)

Depois do `parse()`, `documento['nfeProc']['NFe']['infNFe']` é o "miolo" da nota. A partir dali:

| Caminho no dicionário 🔑 | Informação 🔓 |
|---|---|
| `infNFe['total']['ICMSTot']['vNF']` | valor total da nota |
| `infNFe['emit']['CNPJ']` / `['xNome']` | CNPJ e nome de quem vendeu |
| `infNFe['dest']['CPF']` / `['xNome']` | CPF e nome de quem comprou |
| `infNFe['det']` | lista de produtos — cada item tem `produto['prod']['xProd']` (nome) e `produto['prod']['vProd']` (valor) |

## 🧰 4. Função para Ler uma NFS-e de Serviço

Já a NFS-e (nota de serviço) tem outro caminho, bem mais longo, até chegar nos dados:

| Caminho no dicionário 🔑 | Informação 🔓 |
|---|---|
| `InfNfse['Servico']['Valores']['ValorServicos']` | valor total do serviço |
| `InfNfse['PrestadorServico']['IdentificacaoPrestador']['Cnpj']` / `['RazaoSocial']` (em `PrestadorServico`) | CNPJ e nome de quem prestou o serviço |
| `InfNfse['TomadorServico']['IdentificacaoTomador']['CpfCnpj']['Cnpj']` / `['RazaoSocial']` (em `TomadorServico`) | CPF/CNPJ e nome de quem contratou |
| `InfNfse['Servico']['Discriminacao']` | descrição do serviço prestado |

> ⚠️ **Atenção:** aqui não tem uma lista de "produtos" — é só uma descrição (`Discriminacao`) do serviço.

## 🎯 5. Mentoria: Lendo Todas as Notas da Pasta .spec

Nessa mentoria, juntamos as duas funções (`ler_xml_danfe` e `ler_xml_servico`) pra ler, de uma vez só, todos os XMLs que estão na subpasta `.spec` deste módulo — sem precisar abrir nota por nota manualmente.

**A lógica:**
- Listar os arquivos da pasta `.spec` com `os.listdir()`
- Filtrar só os que terminam em `.xml` (a pasta também tem os PDFs e os `.txt` de outras aulas deste módulo)
- Se o nome do arquivo tiver `"DANFE"`, é nota de produto → usa `ler_xml_danfe`
- Senão, é nota de serviço → usa `ler_xml_servico`

| Arquivo lido 🔑 | Função usada 🔓 |
|---|---|
| `DANFEBrota.xml` | `ler_xml_danfe` |
| `DANFENespresso.xml` | `ler_xml_danfe` |
| `NotaCariocaHashtag.xml` | `ler_xml_servico` |

> ✅ No final, assim como fizemos no desafio da aula anterior, escrevemos um resumo de todas as notas lidas em um novo arquivo txt (`Resultado Notas Fiscais.txt`) na própria subpasta `.spec`.

````python
import os
import xmltodict
from cores import *

def ler_xml_danfe(nota_fiscal):
    with open(f"{nota_fiscal}", 'rb') as arquivo_xml: 
        documento = xmltodict.parse(arquivo_xml)
    dic_notafiscal = documento['nfeProc']['NFe']['infNFe']

    valor_total = dic_notafiscal['total']['ICMSTot']['vNF']
    cnpj_vendeu = dic_notafiscal['emit']['CNPJ']
    nome_vendeu = dic_notafiscal['emit']['xNome']
    cpf_comprou = dic_notafiscal['dest']['CPF']
    nome_comprou = dic_notafiscal['dest']['xNome']
    produtos = dic_notafiscal['det']

    lista_produtos = list()

    for indice_produto, produto in enumerate(produtos):
        valor_produto = produto['prod']['vProd']
        nome_produto = produto['prod']['xProd']
        lista_produtos.append((nome_produto, valor_produto))

    return {
        'tipo': 'NF-e (produto)',
        'valor_total': valor_total,
        'cnpj_vendeu': cnpj_vendeu,
        'nome_vendeu': nome_vendeu,
        'cpf_comprou': cpf_comprou,
        'nome_comprou': nome_comprou,
        'itens': lista_produtos,
    }


def ler_xml_servico(nota_fiscal):
    with open(f"{nota_fiscal}", 'rb') as arquivo_xml:
        documento = xmltodict.parse(arquivo_xml)
    dic_notafiscal = documento['ConsultarNfseResposta']['ListaNfse']['CompNfse']['Nfse']['InfNfse']

    valor_total = dic_notafiscal['Servico']['Valores']['ValorServicos']
    cnpj_vendeu = dic_notafiscal['PrestadorServico']['IdentificacaoPrestador']['Cnpj']
    nome_vendeu = dic_notafiscal['PrestadorServico']['RazaoSocial']
    cpf_comprou = dic_notafiscal['TomadorServico']['IdentificacaoTomador']['CpfCnpj']['Cnpj']
    nome_comprou = dic_notafiscal['TomadorServico']['RazaoSocial']
    servico = dic_notafiscal['Servico']['Discriminacao'].strip()

    return {
        'tipo': 'NFS-e (serviço)',
        'valor_total': valor_total,
        'cnpj_vendeu': cnpj_vendeu,
        'nome_vendeu': nome_vendeu,
        'cpf_comprou': cpf_comprou,
        'nome_comprou': nome_comprou,
        'itens': [(servico, valor_total)],
    }


# a pasta .spec é irmã da pasta .ipynb
pasta_notas = r"../.spec"

notas_lidas = list()

for indice_arquivo, arquivo in enumerate(sorted(os.listdir(pasta_notas))):
    if arquivo.endswith(".xml"):
        caminho_arquivo = os.path.join(pasta_notas, arquivo)
        # só as notas com "DANFE" no nome são NF-e de produto -> 
        # as demais são NFS-e de serviço
        if ("DANFE" in arquivo):
            dados = ler_xml_danfe(caminho_arquivo)
        else:
            dados = ler_xml_servico(caminho_arquivo)
        notas_lidas.append((arquivo, dados))

valor_total_geral = sum(float(dados['valor_total']) for arquivo, dados in notas_lidas)

for (arquivo, dados) in notas_lidas:
    print(f"\n{CinzaClaro}Arquivo:{Reset} {AzulClaro}{arquivo}{Reset} {CinzaEscuro}({dados['tipo']}){Reset}")
    print(f"{CinzaClaro} Vendeu:{Reset} {dados['nome_vendeu']} {CinzaEscuro}(CNPJ {dados['cnpj_vendeu']}){Reset}")
    print(f"{CinzaClaro} Comprou:{Reset} {dados['nome_comprou']} {CinzaEscuro}(CPF/CNPJ {dados['cpf_comprou']}){Reset}")
    print(f"{CinzaClaro} Valor total:{Reset} {VerdeClaro}R${Reset} {CinzaClaro}{dados['valor_total']}{Reset}")
    for nome_item, valor_item in dados['itens']:
        print(f"{CinzaClaro} - {nome_item}:{Reset} R$ {valor_item}")
    print()

print(f"{CinzaClaro}Total geral lido de {len(notas_lidas)} notas:{Reset} {VerdeClaro}R${Reset} {CinzaClaro}{valor_total_geral:.2f}{Reset}")

# escrever um resumo em um novo arquivo txt, junto às notas originais na subpasta .spec
with open(r"../.spec/Resultado Notas Fiscais.txt", "w", encoding="latin-1") as arquivo_xml:
    for (nome_arquivo, dados) in notas_lidas:
        arquivo_xml.write(f"{nome_arquivo} ({dados['tipo']})\n")
        arquivo_xml.write(f"  Vendeu: {dados['nome_vendeu']} (CNPJ {dados['cnpj_vendeu']})\n")
        arquivo_xml.write(f"  Comprou: {dados['nome_comprou']} (CPF/CNPJ {dados['cpf_comprou']})\n")
        arquivo_xml.write(f" Valor total: R$ {dados['valor_total']}\n\n")
    arquivo_xml.write(f"Total geral lido de {len(notas_lidas)} notas: R$ {valor_total_geral:.2f}\n")
````
