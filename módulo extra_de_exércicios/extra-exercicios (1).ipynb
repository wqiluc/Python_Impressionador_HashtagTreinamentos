### Terceira vers√£o da lista de compras

Mantenha o programa da lista de compras com dicion√°rio, mas agora use fun√ß√µes para organizar o c√≥digo. Crie fun√ß√µes para cada uma das op√ß√µes do menu: `adicionar_item`, `remover_item` e `ver_lista`. Crie tamb√©m uma fun√ß√£o para mostrar o menu. O programa deve continuar funcionando da mesma forma, mas agora o c√≥digo deve estar organizado em fun√ß√µes.


````python
from cores import *

dicionario_produtos = dict()

def mostrar_menu():
    opcao = str(input(f"""{Negrito}Digite uma opc√£o do menu:
                      1 - Adicionar Item‚úÖ;
                      2 - Remover Item‚ùå;
                      3 - Vizualizar Lista Completa ‚ö†Ô∏èüëÅÔ∏è; e
                      4 - Sair do Sistema üõíüëã{Reset}""")).strip()
    while not opcao.isdigit() or int(opcao) not in [1,2,3,4]:
        print(f"{Vermelho}Termo Inv√°lido!!‚ùå Digite um N√öMERO entre 1 e 4{Reset}")
        opcao = str(input(f"""{Negrito}Digite uma opc√£o do menu:
                      1 - Adicionar Item‚úÖ;
                      2 - Remover Item‚ùå;
                      3 - Vizualizar Lista Completa ‚ö†Ô∏èüëÅÔ∏è; e
                      4 - Sair do Sistema üõíüëã{Reset}""")).strip()
    return int(opcao)

def adicionar_item():
    adicionar_produto = str(input(f"{Negrito}Digite o nome de um produto: {Reset}")).strip().capitalize()
    while not adicionar_produto.isalpha():
        print(f"{Vermelho}Termo Inv√°lido!!‚ùå Digite um NOME DE PRODUTO (str){Reset}")
        adicionar_produto = str(input(f"{Negrito}Digite o nome de um produto: {Reset}")).strip().capitalize()
    quantidade_produto = str(input(f"{Negrito}Digite quantos(as): '{adicionar_produto}' deseja adicionar?{Reset}")).strip()
    while not quantidade_produto.isdigit():
        print(f"{Vermelho}Termo Inv√°lido!!‚ùå Digite uma QUANTIDADE (int){Reset}")
        quantidade_produto = str(input(f"{Negrito}Digite quantos(as): '{adicionar_produto}' deseja adicionar?{Reset}")).strip()
    dicionario_produtos[adicionar_produto] = int(quantidade_produto)

def remover_item():
    remover_produto = str(input(f"{Negrito}Digite o nome do Produto que quer remover: {Reset}")).strip().capitalize()
    if remover_produto not in dicionario_produtos:
        print(f"{Vermelho}O produto: '{remover_produto}' N√ÉO ENCONTRA-SE no dicion√°rio. Digite novamente{Reset}")
        return
    dicionario_produtos.pop(remover_produto)

def ver_lista():
    for indice_produto, (produto,qtd) in enumerate(dicionario_produtos.items()):
        print(f"{Negrito}{indice_produto+1}¬∫ Produto: {produto} - Quantidade: {qtd}{Reset}")

while True:
    opcao = mostrar_menu()
    if opcao == 1:
        adicionar_item()
    elif opcao == 2:
        remover_item()
    elif opcao == 3:
        ver_lista()
    elif opcao == 4:
        break

print(f"{Verde}Sistema encerrado com sucesso! üõí‚úÖ{Reset} {AzulClaro}Volte sempre!üëã{Reset}")
print(dicionario_produtos)
````
