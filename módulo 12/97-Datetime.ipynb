## Criação de uma data a partir dos valores

Podemos criar um objeto datetime usando o `datetime`. Você pode passar:

- `year`: ano (por exemplo, 2023)
- `month`: mês (1-12)
- `day`: dia (1-31)
- `hour`: hora (0-23)
- `minute`: minuto (0-59)
- `second`: segundo (0-59)
- `microsecond`: microssegundo (0-999999)
- `tzinfo`: fuso horário

```python
ano = (datetime.datetime.now().year)
mes = (datetime.datetime.now().month)
dia = (datetime.datetime.now().day)
minuto = (datetime.datetime.now())
segundo = (datetime.datetime.now().second)

print(f"{CinzaClaro}Data: {dia}/{mes}/{ano} | {minuto}.{segundo} {Reset}")
```

## datetime.timedelta()

A classe `timedelta` é usada para realizar operações com datas (adição e subtração).

```python
from datetime import datetime, timedelta

hoje = datetime.now()
futuro = hoje + timedelta(days=10)

print("Hoje:", hoje)
print("Daqui a 10 dias:", futuro)
```

## Calcular a diferença entre duas datas

Podemos calcular a diferença entre duas datas subtraindo uma da outra. O resultado será um objeto timedelta.


```python
from datetime import datetime

data1 = datetime(2026, 2, 26)
data2 = datetime(2026, 2, 20)

diferenca = data1 - data2

print("Diferença:", diferenca)
print("Dias:", diferenca.days)
```

## Comparação entre datas

Podemos comparar datas usando os operadores de comparação padrão.

Segue uma lógica intuitiva:

    passado < presente < futuro


```python
from datetime import datetime

data1 = str(input("Digite a primeira data (dd/mm/aaaa): "))
data2 = str(input("Digite a segunda data (dd/mm/aaaa): "))

formato = "%d/%m/%Y"

data1 = datetime.strptime(data1, formato)
data2 = datetime.strptime(data2, formato)

if data1 > data2:
    print("A primeira data é mais recente")
elif data1 < data2:
    print("A primeira data é mais antiga")
else:
    print("As datas são iguais")
```
