## datetime.datetime.now()

A função `now()` retorna a data e a hora atuais.


```python
from datetime import datetime
from cores import *

data_atual = datetime.now()
print(data_atual)
```

## Criação de uma data a partir dos valores

Podemos criar um objeto datetime usando o `datetime`. Você pode passar:

- `year`: ano (por exemplo, 2023)
- `month`: mês (1-12)
- `day`: dia (1-31)
- `hour`: hora (0-23)
- `minute`: minuto (0-59)
- `second`: segundo (0-59)
- `microsecond`: microssegundo (0-999999)
- `tzinfo`: fuso horário

```python
from datetime import datetime
#suponde que essas informações estivessem em uma base de dados (.csv, por exemplo)

ano = 2023
mes = 12
dia = 25
minuto = 50
segundo = 9

data_analise = date(ano,mes,dia)
print(data_analise)
```

## datetime.timedelta()

A classe `timedelta` é usada para realizar operações com datas (adição e subtração).

```python
from datetime import datetime, timedelta

hoje = datetime.now()
futuro = hoje + timedelta(days=10)

print("Hoje:", hoje)
print("Daqui a 10 dias:", futuro)
```

## Calcular a diferença entre duas datas

Podemos calcular a diferença entre duas datas subtraindo uma da outra. O resultado será um objeto timedelta.


```python
from datetime import datetime, timedelta

data1 = datetime(2026, 2, 26)
duracao = 365
data2 = data1 + timedelta(days=duracao)

diferenca = data2 - data1

print(f"{CinzaClaro}Diferença:{Reset}",diferenca)
print(f"{CinzaClaro}Dias:{Reset}",diferenca.days)
```

## Comparação entre datas

Podemos comparar datas usando os operadores de comparação padrão.

Segue uma lógica intuitiva:

    passado < presente < futuro


```python
from datetime import datetime,timedelta
from cores import *

data_inicio = datetime(2026, 6, 2)
data_final = data_inicio + timedelta(days=200)
tempo_projeto = (data_final-data_inicio)

print(type(data_final))
print(type(tempo_projeto),"\n")
print(f"{CinzaClaro}O tempo de duração do projeto será de: {tempo_projeto} - Data: {data_final}{Reset}")
```
