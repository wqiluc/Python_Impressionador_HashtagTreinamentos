## 📋 Descrição do Exercício
 
Neste exercício, você irá criar um script Python que:
 
1. **Recebe** a data de nascimento do usuário no formato `dd/mm/aaaa`
2. **Valida** se a data fornecida é válida (dia, mês e ano corretos)
3. **Calcula** a idade exata com base na data atual
4. **Exibe** a idade em anos, meses e dias
---
 
## 🎯 Objetivos de Aprendizado
 
Ao concluir este exercício, você será capaz de:
 
- Utilizar o módulo `datetime` para manipulação de datas
- Converter strings para objetos `date` com `strptime`
- Realizar operações aritméticas entre datas
- Tratar exceções com `while not`
- Validar entradas do usuário com `input()`


```python
from datetime import datetime

idade_usuario = str(input("Digite a sua idade: (dd/mm/aaaa)")).strip()
while (len(idade_usuario) != 10 or (idade_usuario[2]) != "/" or (idade_usuario[5]) != "/"):
    print("Formato de data inválido. ❌ Por favor, digite no formato dd/mm/aaaa.")
    idade_usuario = str(input("Digite a sua idade: (dd/mm/aaaa)")).strip()

dia, mes, ano = idade_usuario.split("/")

dia = int(dia)
mes = int(mes)
ano = int(ano)

if ((dia < 1 or dia > 31) or (mes < 1 or mes > 12) or (ano < 1900 or ano > 2026)):
    print("Datas inválidas. ❌ Por favor, insira uma data válida.")

data_nascimento = datetime(ano, mes, dia)

data_atual = datetime.strptime(datetime.now().strftime("%d/%m/%Y"), "%d/%m/%Y")

idade_formato = data_atual.year - data_nascimento.year - ((data_atual.month, data_atual.day) < (data_nascimento.month, data_nascimento.day))

print(f"Você tem: {idade_formato} anos.")
```
