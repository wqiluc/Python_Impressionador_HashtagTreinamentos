## Tempo de execu√ß√£o

Quanto tempo o c√≥digo abaixo vai demorar para executar? Vamos gerar um n√∫mero para cada um dos funcion√°rios de uma empresa e colocar em uma lista para sortear o funcion√°rio vencedor de um pr√™mio especial de uma viagem de fim de ano. Quanto tempo seu c√≥digo demora para criar essa lista de n√∫meros a serem sorteados?

```python
import time

tempo_inicial = time.time()

sorteio_funcionarios = list()
qtd_funcionarios = 1000000

for indice_f in range(qtd_funcionarios):
    funcionario_sorteado = indice_f
    sorteio_funcionarios.append(funcionario_sorteado)

tempo_final = time.time()

resultado_tempo = (tempo_final-tempo_inicial)

print(f"{CinzaClaro}{resultado_tempo:2f}segundos! {Reset}")
```
