## 1. Input até o usuário parar

Vamos criar um sistema de vendas. Nosso programa deve registrar os produtos e as quantidades (2 inputs) e adicionar em uma lista.

O programa deve continuar rodando até o input ser vazio, ou seja, o usuário apertar enter sem digitar nenhum produto ou quantidade.

Ao final do programa, ele deve printar todos os produtos e quantidades vendidas.

Obs: Caso queira, para o print ficar mais visual, pode usar o join para cada item ser printado em uma linha.
Sugestão para sua lista de produtos vendidos:

```python
from cores import (Negrito, Reset, Vermelho, Verde)

vendas = []

opcao = "S"

while opcao == "S":
    produto = input(f"{Negrito}Digite o nome do Produto: {Reset}").strip().capitalize()
    while not produto.isalpha():
        print(f"{Vermelho}Termo Inválido!! Digite um PRODUTO (str){Reset}")
        produto = str(input(f"{Negrito}Digite o nome do Produto: {Reset}").strip().capitalize())

    preco = input(f"{Negrito}Digite o preço do produto '{produto}': R${Reset}")
    while not preco.isdigit():
        print(f"{Vermelho}Termo Inválido!! Digite o PREÇO do produto (int){Reset}")
        preco = str(input(f"{Negrito}Digite o preço do produto '{produto}': R${Reset}"))

    vendas.append([produto, int(preco)])

    opcao = str(input(f"{Negrito}Deseja Continuar Comprando? [S/N]{Reset}").strip().upper())
    while opcao not in ("SN"):
        print(f"{Vermelho}Termo Inválido!! Digite uma opção [S/N]{Reset}")
        opcao = input(f"{Negrito}Deseja Continuar Comprando? [S/N]{Reset}").strip().upper()

for indice_venda_preco in range(len(vendas)):
    print(f"{Negrito}{indice_venda_preco+1}º Produto: {vendas[indice_venda_preco][0].capitalize()} = {Reset}{Verde}R${Reset}{vendas[indice_venda_preco][1]:.2f}")
```

### Obs: Podemos fazer o While de 2 maneiras:

1. While com a condição que finalize o programa
2. While rodando para sempre, mas com uma condição dentro do while que dê um break no código.
