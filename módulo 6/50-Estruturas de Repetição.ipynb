#### 4. Supondo que a popula√ß√£o de um pa√≠s A seja da ordem de 80000 habitantes com uma taxa anual de crescimento de 3% e que a popula√ß√£o de B seja 200000 habitantes com uma taxa de crescimento de 1.5%.
````python
pa√≠sA = 80000
pa√≠sB = 200000

anos = 0

while pa√≠sA<pa√≠sB:
    pa√≠sA = pa√≠sA + (0.03*pa√≠sA)
    pa√≠sB = pa√≠sB + (0.015*pa√≠sB)
    anos+=1

print(f"{Negrito}{anos} Anos{Reset}")
````
