# For nos Dicionarios

### Estrutura:

```python
from cores import(Negrito,Reset,Verde)
vendas_tecnologia = {'iphone': 15000, 'samsung galaxy': 12000, 'tv samsung': 10000, 'ps5': 14300, 'tablet': 1720, 'ipad': 1000, 'tv philco': 2500, 'notebook hp': 1000, 'notebook dell': 17000, 'notebook asus': 2450}

#demonstrando o for

for indice_produto, vendas_tecnologia in enumerate(vendas_tecnologia.items()):
    print(f"{Negrito}{indice_produto+1}º Produto: {vendas_tecnologia[0].capitalize()} - {Reset}{Verde}R${Reset}{vendas_tecnologia[1]:.2f}{Reset}")
```

- Qual o total de notebooks vendidos?

```python
total_notebook = 0
for indice_notebook in vendas_tecnologia:
    if ("notebook" in indice_notebook):
        total_notebook+=vendas_tecnologia[indice_notebook]

print(f"{Negrito}O total, em Notebooks, foi de:{Reset} {Verde}R${Reset}{Negrito}{total_notebook:.2f}{Reset}")
```
