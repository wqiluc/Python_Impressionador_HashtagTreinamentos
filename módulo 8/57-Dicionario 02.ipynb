### Python Vers√µes Antigas x Vers√µes Novas

- Dicion√°rios eram "sem ordem". Atualmente tem ordem, mas o certo √© usar as chaves
- 2 formas de pegar um valor:
    - dicionario[chave]
    - .get(chave)

```python
from cores import(Negrito,Reset,Verde)
mais_vendidos = {'tecnologia': 'iphone', 'refrigeracao': 'ar consul 12000 btu', 'livros': 'o alquimista', 'eletrodom√©stico': 'geladeira', 'lazer': 'prancha surf'}

vendas_tecnologia = {'iphone': 15000, 'samsung galaxy': 12000, 'tv samsung': 10000, 'ps5': 14300, 'tablet': 1720, 'ipad': 1000, 'tv philco': 2500, 'notebook hp': 1000, 'notebook dell': 17000, 'notebook asus': 2450}
```

- Qual foi o item mais vendido nas categorias 'livros' e 'lazer'?
- Quanto foi vendido de 'notebook asus' e de 'ipad'?

```python
# respondendo com a chave, igual a aula anterior
mais_vendido_livro = mais_vendidos["livros"]
mais_vendido_lazer = mais_vendidos["lazer"]
mais_vendido_tecnologia_asus = vendas_tecnologia["notebook asus"]
mais_vendido_tecnologia_ipad = vendas_tecnologia["ipad"]

print(f"{Negrito}O livro mais vendido foi: {mais_vendido_livro.capitalize():<10}{Reset}")
print(f"{Negrito}O Produto de lazer mais vendido foi: {mais_vendido_lazer.capitalize():<10}{Reset}")
print(f"{Negrito}O Valor do Notebook Asus mais vendido foi de:{Reset} {Verde}R${Reset}{Negrito}{mais_vendido_tecnologia_asus:.2f}{Reset}")
print(f"{Negrito}O Valor do iPad mais vendido foi de:{Reset} {Verde}R${Reset}{Negrito}{mais_vendido_tecnologia_ipad:.2f}{Reset}")

# respondendo com o m√©todo get
mais_vendido_livro = mais_vendidos.get("livros")
mais_vendido_lazer = mais_vendidos.get("lazer")
mais_vendido_tecnologia_asus = vendas_tecnologia.get("notebook asus")
mais_vendido_tecnologia_ipad = vendas_tecnologia.get("ipad")
```
