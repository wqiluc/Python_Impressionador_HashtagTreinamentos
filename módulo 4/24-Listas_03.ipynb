# Como descobrir o índice de um item de uma lista?

i = lista.index('item')

Exemplo:

Digamos que você puxou do Banco de Dados da sua empresa uma lista com todos os produtos que a empresa vende e a quantidade em estoque de todos eles.

```python
produtos = ['tv', 'celular', 'tablet', 'mouse', 'teclado', 'geladeira', 'forno']
estoque = [100, 150, 100, 120, 70, 180, 80]
```

Nesse caso a lista é "pequena" para fins didáticos, mas essa lista poderia ter dezenas de milhares de produtos diferentes.

E agora, como eu faço para descobrir a quantidade em estoque do produto geladeira?

```python
from cores import (Negrito, Reset, Verde)

produtos = ['tv', 'celular', 'tablet', 'mouse', 'teclado', 'geladeira', 'forno']
estoque = [100, 150, 100, 120, 70, 180, 80]

print(f"\n {Negrito}O estoque arrecadou um total de:{Reset} {Verde}R${Reset}{Negrito}{sum(estoque)}{Reset}")

print(f"\n {Negrito}A Tv está na posiçao: {produtos.index('tv')} e ela arrecadou:{Reset} {Verde}R${Reset}{Negrito}{estoque[0]}{Reset}")
print(f"\n {Negrito}O Celular está na posiçao: {produtos.index('celular')} e ele arrecadou:{Reset} {Verde}R${Reset}{Negrito}{estoque[1]}{Reset}")
print(f"\n {Negrito}O Tablet está na posiçao: {produtos.index('tablet')} e ele arrecadou:{Reset} {Verde}R${Reset}{Negrito}{estoque[2]}{Reset}")
print(f"\n {Negrito}O Mouse está na posiçao: {produtos.index('mouse')} e ele arrecadou:{Reset} {Verde}R${Reset}{Negrito}{estoque[3]}{Reset}")
print(f"\n {Negrito}O Teclado está na posiçao: {produtos.index('teclado')} e ele arrecadou:{Reset} {Verde}R${Reset}{Negrito}{estoque[4]}{Reset}")
print(f"\n {Negrito}A Geladeira está na posiçao: {produtos.index('geladeira')} e ela arrecadou:{Reset} {Verde}R${Reset}{Negrito}{estoque[5]}{Reset}")
print(f"\n {Negrito}O Forno está na posiçao: {produtos.index('forno')} e ele arrecadou:{Reset} {Verde}R${Reset}{Negrito}{estoque[6]}{Reset}")
```

Crie um programa para fazer uma consulta de estoque. O usuário do programa deve inserir o nome do produto e, caso ele não exista na lista, ele deve ser avisado. Caso exista, o programa deve dizer a quantidade de unidades em estoque do produto

```python
from cores import (Negrito, Reset, Verde, Vermelho)

produtos = [
    'tv', 'celular', 'tablet', 'mouse', 'teclado', 'geladeira', 'forno', 'monitor',
    'notebook', 'impressora', 'scanner', 'roteador', 'modem', 'webcam', 'microfone',
    'caixa de som', 'headset', 'controle', 'videogame', 'hd externo', 'ssd',
    'pendrive', 'memoria ram', 'placa de video', 'processador', 'fonte',
    'gabinete', 'cooler', 'estabilizador', 'nobreak', 'smartwatch', 'camera',
    'drone', 'projetor', 'tv box', 'repetidor', 'switch', 'adaptador',
    'carregador', 'cabo hdmi', 'cabo usb', 'cabo de rede', 'hub usb',
    'teclado mecanico', 'mouse gamer', 'monitor gamer', 'fone bluetooth',
    'ring light', 'suporte de celular', 'suporte de monitor', 'mesa digitalizadora'
]

produto = input(f"\n{Negrito}Digite o nome do produto para consulta:{Reset} ").lower().strip()

if produto in produtos:
    print(f"\n{Verde}O produto: '{produto}', está na lista ✅ e se encontra na posição: {produtos.index(produto)}{Reset}")
else:
    print(f"\n{Vermelho}{Negrito}O produto: {produto} NÃO SE ENCONTRA NA LISTA ❌{Reset}")


```
