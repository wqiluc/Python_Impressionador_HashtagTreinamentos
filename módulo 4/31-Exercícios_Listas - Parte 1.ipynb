## 1. Faturamento do Melhor e do Pior Mês do Ano

Qual foi o valor de vendas do melhor mês do Ano?
E valor do pior mês do ano?

```python
from cores import(Negrito,Reset,Verde)

meses = ['jan', 'fev', 'mar', 'abr', 'mai', 'jun',  'jul', 'ago', 'set', 'out', 'nov', 'dez']
vendas_1sem = [25000, 29000, 22200, 17750, 15870, 19900]
vendas_2sem = [19850, 20120, 17540, 15555, 49051, 9650]
vendas_ano = vendas_1sem+vendas_2sem

print(f"Meses: {meses}")
print(f"Vendas: {Verde}R${Reset}{vendas_ano}")

print(f"\n {Negrito}O melhor mês do ano teve um valor de:{Reset} {Verde}R${Reset}{max(vendas_ano):.2f}{Reset}")
print(f"\n {Negrito}O pior mês do ano teve um valor de:{Reset} {Verde}R${Reset}{min(vendas_ano):.2f}{Reset}")
```

## 2. Continuação

Agora relacione as duas listas para printar 'O melhor mês do ano foi {} com {} vendas' e o mesmo para o pior mês do ano.

Calcule também o faturamento total do Ano e quanto que o melhor mês representou do faturamento total.

Obs: Para o faturamento total, pode usar a função sum(lista) que soma todos os itens de uma lista

```python
from cores import(Negrito,Reset,Verde,Azul,Vermelho,AmareloClaro)

meses = ['jan', 'fev', 'mar', 'abr', 'mai', 'jun',  'jul', 'ago', 'set', 'out', 'nov', 'dez']
vendas_1sem = [25000, 29000, 22200, 17750, 15870, 19900]
vendas_2sem = [19850, 20120, 17540, 15555, 49051, 9650]

vendas_ano = (vendas_1sem+vendas_2sem)

print(f"Meses: {meses}")
print(f"Vendas: {Verde}R${Reset}{vendas_ano}")

print(f"\n {Negrito}O melhor mês do ano foi:{Reset} {Azul}{meses[10]}{Reset}{Negrito}, que teve um valor de:{Reset} {Verde}R${Reset}{max(vendas_ano):.2f}{Reset}")
print(f"\n {Negrito}O pior mês do ano foi:{Reset} {Vermelho}{meses[11]}{Reset}{Negrito}, que teve um valor de:{Reset} {Verde}R${Reset}{min(vendas_ano):.2f}{Reset}")
print(f"\n {Negrito}O{Reset} {AmareloClaro}faturamento total{Reset} desse ano foi de:{Reset} {Verde}R${Reset}{sum(vendas_ano):.2f}{Reset}")
```

## 3. Crie uma lista com o top 3 valores de vendas do ano (sem fazer "no olho")

Dica: o método remove retira um item da lista.

```python
top3 = [ ]

for _ in range(3):
    máximo_valor = max(vendas_ano)
    top3.append(máximo_valor)
    vendas_ano.remove(máximo_valor)
print(f"\n {Negrito}Os 3 melhores meses do ano foram:{top3}, \nque tiveram um valor de:{Reset} {Verde}R${Reset}{sum(top3)}{Reset}")
```
