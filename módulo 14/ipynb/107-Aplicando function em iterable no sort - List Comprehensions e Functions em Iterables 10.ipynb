## 💡 1. Contexto — o problema da ordenação padrão

Até agora no programa usamos várias vezes o `.sort()` para ordenar listas. Só que o Python ordena strings pela tabela Unicode — e nela, **todas as letras maiúsculas vêm antes das minúsculas**. Isso faz a ordenação "parecer" errada quando a lista mistura maiúsculas e minúsculas.

| Caractere | Valor Unicode |
|-----------|---------------|
| `'A'` ... `'Z'` | 65 – 90 |
| `'a'` ... `'z'` | 97 – 122 |

> ⚠️ **Cuidado:** por isso `'IPad'` aparece antes de `'airpods'` num sort padrão — o `I` maiúsculo "vale menos" que o `a` minúsculo para o Python.

---

## 🔑 2. `sort(key=...)` — ordenando do seu jeito

O parâmetro `key` recebe uma **função** que é aplicada a cada item antes da comparação. O item original não muda — só o critério usado para comparar muda.

**Sintaxe:**
```
lista.sort(key=função)
sorted(iteravel, key=função)
```

| `key=...` | O que faz |
|-----------|-----------|
| 🔡 `str.lower` | Ignora maiúsculas/minúsculas na comparação |
| 📏 `len` | Ordena pelo tamanho do item |
| ⚡ `lambda x: ...` | Ordena por um critério customizado |

> 💡 **Dica:** `str.lower` funciona porque o Python chama `str.lower(item)` para cada item — não precisa criar uma função nova quando uma já pronta resolve.

---

## ⚖️ 3. `sort()` vs `sorted()`

| | `.sort()` | `sorted()` |
|---|-----------|------------|
| Onde funciona | Só em `list` | Qualquer iterável (lista, dict, tupla, string...) |
| Retorno | `None` (ordena *in-place*) | Uma **nova** lista |
| Original é alterado? | ✅ Sim | ❌ Não |

> ⚠️ **Cuidado:** dicionários não têm `.sort()` — para ordenar um dict, use `sorted()`.

---

## 📖 4. Ordenando dicionários por valor

Como dicionários não têm método próprio de ordenação, usamos `sorted()` sobre `.items()` (pares chave-valor) com uma `lambda` que escolhe o valor como critério.

```python
vendas_produtos = {'vinho': 100, 'cafeiteira': 150, 'microondas': 300, 'iphone': 5500}

lista_vendas = sorted(vendas_produtos.items(), key=lambda item: item[1], reverse=True)
# → [('iphone', 5500), ('microondas', 300), ('cafeiteira', 150), ('vinho', 100)]

dict(lista_vendas)
# → {'iphone': 5500, 'microondas': 300, 'cafeiteira': 150, 'vinho': 100}
```

| Parte da lambda | Significado |
|------------------|-------------|
| `item` | Uma tupla `(chave, valor)` vinda de `.items()` |
| `item[1]` | O valor — é isso que vira o critério de ordenação |
| `reverse=True` | Ordena do maior para o menor |

---

## 🤔 5. Quando usar cada abordagem?

| Situação | Use |
|----------|-----|
| Lista simples, ignorando maiúsculas/minúsculas | `.sort(key=str.lower)` |
| Precisa manter a lista original intacta | `sorted(lista)` |
| Ordenar dicionário por valor | `sorted(dict.items(), key=lambda item: item[1])` |
| Critério de ordenação complexo | `lambda` customizada em `key` |

````python
produtos = ['apple tv', 'mac', 'IPhone x', 'IPhone 11', 'IPad', 'apple watch', 'mac book', 'airpods']
produtos.sort()
print(produtos)
````

- Como faríamos para ordenar corretamente?

````python
produtos.sort(key=str.lower)
print(produtos)
````

### Outro exemplo: como ordenar um dicionário de acordo com o valor

````python
vendas_produtos = {'vinho': 100, 'cafeiteira': 150, 'microondas': 300, 'iphone': 5500}
````

- Queremos listar da maior quantidade de vendas para a menor, para enviar como report para o diretor, por exemplo

````python
lista_vendas = sorted(vendas_produtos.items(), key=lambda item: item[1], reverse=True)

print(dict(lista_vendas))
````
