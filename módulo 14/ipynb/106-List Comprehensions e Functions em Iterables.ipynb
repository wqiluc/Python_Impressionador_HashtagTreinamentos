## 💡 1. Contexto — por que isso existe?
 
Quando você aprendeu **list comprehension**, viu que dá pra aplicar uma lógica a cada item de uma lista sem usar um loop explícito. O Python também tem funções built-in que fazem isso de forma ainda mais direta — e são a base da **programação funcional**.
 
| Função | O que faz |
|--------|-----------|
| 🔄 `map()` | Transforma cada item com uma função |
| 🔍 `filter()` | Mantém só os itens que passam em um teste |
| ➕ `reduce()` | Acumula todos os itens em um valor único |
| ⚡ `lambda` | Funções anônimas, parceiras do map/filter |
 
---
 
## 🗺️ 2. `map()` — transformar cada item
 
A `map()` aplica uma função a cada elemento de um iterável e retorna um objeto `map` (que você converte com `list()`).
 
**Sintaxe:**
```
map(função, iterável)
```
 
- **Retorna:** objeto map (lazy — converta com `list()`)
- **Vários iteráveis:** `map(fn, lista1, lista2)` → `fn` recebe dois args
```python
# Exemplo básico — dobrar cada número
numeros = [1, 2, 3, 4, 5]
 
def dobrar(n):
    return n * 2
 
resultado = list(map(dobrar, numeros))
# → [2, 4, 6, 8, 10]
```
 
```python
# Com lambda — versão compacta
resultado = list(map(lambda n: n * 2, numeros))
# → [2, 4, 6, 8, 10]
```
 
```python
# map com strings
nomes = ["ana", "bia", "carlos"]
capitalizados = list(map(str.capitalize, nomes))
# → ["Ana", "Bia", "Carlos"]
```
 
> 💡 **Dica:** quando a função já existe (como `str.upper`, `int`, `float`), você passa ela diretamente sem criar um lambda.

## 🔍 3. `filter()` — filtrar por condição
 
A `filter()` mantém apenas os itens para os quais a função retorna `True`.
 
```python
# Manter só números pares
numeros = [1, 2, 3, 4, 5, 6]
 
pares = list(filter(lambda n: n % 2 == 0, numeros))
# → [2, 4, 6]
 
# Manter só strings não-vazias
palavras = ["oi", "", "python", "", "legal"]
validas = list(filter(None, palavras))
# → ["oi", "python", "legal"]
```
 
> ⚠️ **Cuidado:** `filter(None, iterable)` remove qualquer valor *falsy* (`0`, `""`, `None`, `[]`, `False`). Isso pode surpreender se você quiser manter zeros.

## ⚖️ 4. Comparando as 3 abordagens
 
Para dobrar os números pares de uma lista — três formas de escrever a mesma lógica:
 
```python
numeros = [1, 2, 3, 4, 5, 6]
 
# ① For loop clássico
resultado = []
for n in numeros:
    if n % 2 == 0:
        resultado.append(n * 2)
 
# ② List comprehension
resultado = [n * 2 for n in numeros if n % 2 == 0]
 
# ③ map + filter
resultado = list(map(lambda n: n * 2, filter(lambda n: n % 2 == 0, numeros)))
 
# Todas produzem → [4, 8, 12]
```
 
> 💡 As três produzem o mesmo resultado. A diferença está na legibilidade e no estilo.

## 🤔 5. Quando usar cada abordagem?
 
| Abordagem | Use quando... |
|-----------|---------------|
| 🔁 **For loop** | A lógica é complexa ou você precisa de efeitos colaterais (`print`, `append` em estruturas externas) |
| 📋 **List comprehension** | Quer transformar e/ou filtrar em uma linha legível. A mais *pythônica* para a maioria dos casos |
| 🗺️ **map / filter** | A função já existe (`str.upper`, `int`) ou você está compondo pipelines de transformação |

```python
lista = list(map(função, iterable_original))
```

- Exemplo: digamos que eu tenha uma function que corrige um código de um produto (semelhante ao que fizemos na seção de function aqui do curso)

````python
def padronizar_texto(texto):
    texto = texto.casefold()
    texto = texto.replace("  ", " ")
    texto = texto.strip()
    return texto
````

- Agora queremos padronizar uma lista de códigos:

````python
produtos = [' ABC12 ', 'abc34', 'AbC37', 'beb12', ' BSA151', 'BEB23']
````

- Usando o for, temos que percorrer a lista toda e para cada item executar a function

````python

print(produtos)
````

- Usando o map, apenas chamamos a função e ela já faz isso para a gente

````python
print(produtos)
````
