## 💡 1. Contexto — lambda como argumento

Até agora criamos lambdas e guardamos elas em uma variável. Mas a **aplicação mais comum** é usar a lambda direto onde uma function é esperada — como argumento de `map()` e `filter()` — sem precisar dar nome a ela.

```python
preco_tecnologia = {
    'notebook asus': 2450, 'iphone': 4500, 'samsung galaxy': 3000,
    'tv samsung': 1000, 'ps5': 3000, 'tablet': 1000,
    'notebook dell': 3000, 'ipad': 3000, 'tv philco': 800, 'notebook hp': 1700
}
```

---

## 🗺️ 2. `map()` — function vs lambda

**Objetivo:** somar 30% de imposto ao preço de cada produto.

```python
# Fazendo por function
def calcular_imposto(preco):
    return preco + (preco * 0.3)

precos_com_imposto = list(map(calcular_imposto, preco_tecnologia.values()))
```

```python
# Fazendo por lambda — sem precisar nomear a function
precos_com_imposto = list(map(lambda preco: preco + (preco * 0.3), preco_tecnologia.values()))
```

| Abordagem | Vantagem |
|-----------|----------|
| `def` | Reaproveitável em outros lugares do código |
| `lambda` | Direto no local de uso, sem "poluir" o namespace com uma function que só serve ali |

---

## 🔍 3. `filter()` — function vs lambda

**Objetivo:** manter só os produtos que custam acima de 2000.

```
filter(função, iterable) -> retorna os itens do iterable onde a função é True
```

```python
# Fazendo por function
def acima_de_2000(item):
    return item[1] > 2000

produtos_caros = list(filter(acima_de_2000, preco_tecnologia.items()))
```

```python
# Fazendo por lambda
produtos_caros = list(filter(lambda item: item[1] > 2000, preco_tecnologia.items()))
```

> 💡 **Dica:** ao filtrar um dicionário, usamos `.items()` — a função recebe uma tupla `(chave, valor)`, por isso `item[1]` acessa o valor.

---

## ⚖️ 4. Comparando as duas abordagens

| | `def` (function nomeada) | `lambda` |
|---|---------------------------|----------|
| Precisa de nome | ✅ Sim | ❌ Não |
| Ideal para | Lógica reutilizável em vários lugares | Lógica de uso único, direto no `map`/`filter` |
| Legibilidade | Melhor para lógica complexa | Melhor para lógica curta e simples |

> 💡 O resultado final é **idêntico** nas duas formas — a escolha é sobre legibilidade e reutilização, não sobre capacidade.

````python
preco_tecnologia = {'notebook asus': 2450, 'iphone': 4500, 'samsung galaxy': 3000, 'tv samsung': 1000, 'ps5': 3000, 'tablet': 1000, 'notebook dell': 3000, 'ipad': 3000, 'tv philco': 800, 'notebook hp': 1700}
````

### map()
- Queremos saber o preço de cada produto adicionando o valor do imposto de 30% sobre o valor do produto

````python
#Fazendo por function
def calcular_imposto(preco):
    return preco + (preco * 0.3)

precos_com_imposto = list(map(calcular_imposto, preco_tecnologia.values()))
print(precos_com_imposto)
````

````python
#fazendo por lambda
precos_com_imposto = list(map(lambda preco: preco + (preco * 0.3), preco_tecnologia.values()))
print(precos_com_imposto)
````

### filter()
- Queremos apenas os produtos que custam acima de 2000

````python
#fazendo por function
#print(preco_tecnologia.items())
def acima_de_2000(item):
    return item[1] > 2000

produtos_caros = list(filter(acima_de_2000, preco_tecnologia.items()))
print(produtos_caros)
````

````python
#fazendo por lambda
produtos_caros = list(filter(lambda item: item[1] > 2000, preco_tecnologia.items()))
print(produtos_caros)
````
