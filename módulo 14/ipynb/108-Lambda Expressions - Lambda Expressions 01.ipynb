## 💡 1. O que é uma lambda expression?

Uma **lambda expression** é uma função anônima (sem nome) de **uma única linha**, atribuída a uma variável — a variável passa a se comportar como a função.

| Característica | Function normal (`def`) | `lambda` |
|-----------------|--------------------------|----------|
| Nome | Tem (`def minha_funcao`) | Não tem (anônima) |
| Corpo | Múltiplas linhas, `return` explícito | Uma linha, retorno implícito |
| Uso típico | Lógica reaproveitável, complexa | Ação única, rápida, geralmente como argumento |

> 💡 **Dica:** quase tudo que se faz com `lambda` também dá pra fazer com uma function normal — a diferença é a conveniência de não precisar nomear e definir uma function só para uma linha de lógica.

### 🔎 Principais aplicações
- Como **argumento** dentro de métodos (`sort`, `map`, `filter`...) — sem precisar criar uma function separada (próxima aula)
- Como **gerador de funções** — uma function que retorna outra function (aula 3 desta seção)

---

## ⚙️ 2. Estrutura

```
minha_funcao = lambda parametro: expressão
```

| Parte | Significado |
|-------|-------------|
| `minha_funcao` | Nome da variável que guarda a lambda |
| `lambda` | Palavra-chave que inicia a expressão |
| `parametro` | Um ou mais parâmetros (igual em uma function normal) |
| `expressão` | O que a função calcula e retorna — sem precisar de `return` |

---

## 🔢 3. Exemplo mais simples

Uma lambda que dobra um número:

```python
dobro = lambda numero: numero * 2
print(dobro(5))
# → 10
```

---

## 💰 4. Exemplo útil — preço com imposto

Uma lambda que calcula o preço de um produto já com o imposto embutido:

```python
imposto = 0.3
preco_com_imposto = lambda preco: preco + (preco * imposto)

print(preco_com_imposto(100))
# → 130.0
```

> ⚠️ **Cuidado:** a lambda "enxerga" variáveis de fora (como `imposto`) no momento em que é **chamada**, não em que é criada — se `imposto` mudar depois, o resultado muda junto.

---

## 🤔 5. Quando usar lambda?

| Situação | Use |
|----------|-----|
| Lógica de uma linha, usada uma vez | `lambda` |
| Vai passar a lógica como argumento (`key=`, `map`, `filter`) | `lambda` |
| Lógica com várias linhas, condicionais, loops | `def` (function normal) |
| Vai reutilizar a mesma lógica em vários lugares com nome próprio | `def` (function normal) |

- Exemplo mais simples

````python
dobro = lambda numero: numero * 2

print(dobro(5))
````

- Exemplo útil: Vamos usar lambda expressions para criar uma função que calcula o preço dos produtos acrescido do imposto

````python
imposto = 0.3

preco_com_imposto = lambda preco: preco + (preco * imposto)

print(preco_com_imposto(100))
````
