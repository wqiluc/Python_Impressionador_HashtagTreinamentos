## 💡 1. Contexto — o problema

Queremos calcular o valor de uma nota fiscal acrescido do imposto — mas cada **categoria** tem uma alíquota diferente:

| Categoria | Alíquota |
|-----------|----------|
| 📦 Produto | 10% |
| 🛠️ Serviço | 15% |
| 💰 Royalties | 25% |

Poderíamos criar 3 functions praticamente idênticas, uma para cada alíquota. Mas isso repete a mesma lógica 3 vezes — só o número muda.

---

## 🏭 2. O gerador de funções

A solução: uma function que **recebe a alíquota** e **retorna uma lambda** já configurada com esse valor.

```python
def gerador_de_imposto(taxa):
    return lambda valor: valor + (valor * taxa)

calcular_imposto_produto = gerador_de_imposto(0.1)
calcular_imposto_servico = gerador_de_imposto(0.15)
calcular_imposto_royalties = gerador_de_imposto(0.25)
```

| Parte | Significado |
|-------|-------------|
| `gerador_de_imposto(taxa)` | A "fábrica" — recebe a alíquota e devolve uma function pronta |
| `lambda valor: ...` | A function "fabricada" — já guarda a `taxa` que recebeu |
| `calcular_imposto_produto` | Uma function completa, pronta para calcular imposto de 10% |

> 💡 **Dica:** cada função gerada guarda sua própria `taxa` — é isso que chamamos de *closure*: a lambda "lembra" do valor que existia no momento em que foi criada.

---

## 🔢 3. Aplicando as funções geradas

```python
nota_fiscal = 100

print(calcular_imposto_produto(nota_fiscal))     # → 110.0
print(calcular_imposto_servico(nota_fiscal))      # → 115.0
print(calcular_imposto_royalties(nota_fiscal))    # → 125.0
```

Cada função criada por `gerador_de_imposto` se comporta como uma function independente, mas todas compartilham a mesma lógica interna.

---

## 🤔 4. Quando vale a pena?

| Situação | Vale a pena? |
|----------|--------------|
| Mesma lógica, aplicada com parâmetros fixos diferentes (taxas, multiplicadores...) | ✅ Sim — evita repetir a function várias vezes |
| Lógica simples, usada uma única vez | ❌ Não — um `lambda` direto já resolve |
| Lógica com múltiplas linhas ou efeitos colaterais | ❌ Não — prefira `def` normal |

> ⚠️ **Cuidado:** essa técnica é menos usada no dia a dia que `lambda` como argumento de `map`/`filter`, mas é útil quando você precisa "customizar" funções em massa.

````python
# produto 0.1
#serviço 0.15
#royalties 0.25
````

- Agora vamos definir as funções que calculam o imposto das 3 categorias (produto, serviço, royalties)

````python
def gerador_de_imposto(taxa):
    return lambda valor: valor + (valor * taxa)

calcular_imposto_produto = gerador_de_imposto(0.1)
calcular_imposto_servico = gerador_de_imposto(0.15)
calcular_imposto_royalties = gerador_de_imposto(0.25)
````

- Agora vamos aplicar com um valor de nota fiscal de 100 para ver o resultado

````python
nota_fiscal = 100

print(calcular_imposto_produto(nota_fiscal))
print(calcular_imposto_servico(nota_fiscal))
print(calcular_imposto_royalties(nota_fiscal))
````
