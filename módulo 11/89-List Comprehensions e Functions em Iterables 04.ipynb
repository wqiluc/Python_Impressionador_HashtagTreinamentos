# List Comprehensions com if para "filtrar" itens

### Estrutura:

- Digamos que eu queira criar uma lista de produtos que bateram a meta.

```python
from cores import *
meta = 1000
vendas_produtos_lista = [1500, 150, 2100, 1950]
produtos_lista = ['vinho', 'cafeiteira', 'microondas', 'iphone']
```

Fazendo por For tradicional

```python
for indice_produto, (vendas_acima,produto_acima) in enumerate(zip(vendas_produtos_lista,produtos_lista)):
    if (vendas_acima>meta):
        print(f"{Negrito}{indice_produto+1}º Produto: {produto_acima.capitalize():<5} - Preço:{Reset} {Verde}R${Reset}{Negrito}{vendas_acima:.2f}{Reset}")
    else:
        pass
```

Fazendo por list comprehension

```python
produto_acima_da_meta = list(produto for indice_produto, produto in enumerate(produtos_lista) if vendas_acima>meta)
produto_acima_da_meta.remove("cafeiteira")
print(produto_acima_da_meta)
```
