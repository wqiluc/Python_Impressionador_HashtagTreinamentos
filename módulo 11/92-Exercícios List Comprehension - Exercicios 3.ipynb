## 1. Tamanho do Pedido de Compras üõíüõçÔ∏è
<strong>Nesse exerc√≠cio vamos avaliar o estoque de uma empresa. Vamos considerar que todos os produtos dessa empresa s√£o comprados em lotes de 500 unidades.</strong>

- Caso o estoque esteja abaixo de 1000 unidades, devemos fazer um pedido de 500 unidades.
- Caso o estoque esteja abaixo de 200 unidades, devemos fazer um pedido de 1000 unidades.

<strong>Defina o valor a ser pedido de cada produto para enviar ao time de compras.</strong>

```python
from cores import *
estoque = [('BSA2199',396),('PPF5239',251),('BSA1212',989),('PPF2154',449),('BEB3410',241),('PPF8999',527),('EMB9591',601),('BSA2006',314),('EMB3604',469),('EMB2070',733),('PPF9018',339),('PPF1468',906),('BSA5819',291),('PPF8666',850),('BEB2983',353),('BEB5877',456),('PPF5008',963),('PPF3877',185),('PPF7321',163),('BSA8833',644),('PPF4980',421),('PPF3063',757),('BSA2089',271),('BSA8398',180),('EMB4622',515),('EMB9814',563),('PPF3784',229),('PPF2398',270),('BEB3211',181),('PPF8655',459),('PPF1874',799),('PPF8789',126),('PPF6324',375),('EMB9290',883),('BSA5516',555),('BSA8451',243),('BSA8213',423)]
```

```python
#1¬∫ Fazendo por for:
lista_1000 = []
lista_200 = []

for indice_estoque, (nome,qtd) in enumerate(estoque):
    if (qtd <= 200):
        print(f"{Amarelo}{qtd}+1000! ‚úÖ{Reset}")
        qtd += 1000
        lista_200.append([nome,qtd])
    elif (qtd <= 1000):
        print(f"{Verde}{qtd}+500! ‚úÖ{Reset}")
        qtd += 500
        lista_1000.append([nome,qtd])
    print(f"{Negrito}{indice_estoque+1}¬∫ Produto --> Nome: '{nome}' - Quantidade: {qtd}{Reset}")

print("\n",lista_1000)
print(lista_200)
```

```python
#2¬∫ Fazendo por List Comprehension:

lista_200 = list((nome, qtd+1000) for (nome,qtd) in estoque if qtd <= 200)

lista_1000 = list((nome, qtd+500) for (nome,qtd) in estoque if 200 < qtd <= 1000)
```
