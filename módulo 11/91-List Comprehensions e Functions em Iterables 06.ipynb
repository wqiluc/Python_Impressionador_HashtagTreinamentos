# List Comprehension com if para escolher o resultado final

### Estrutura:

- Digamos que eu esteja analisando os vendedores de uma loja e queira criar uma lista para enviar para o RH com o bônus de cada vendedor.
- O bônus é dado por 10% do valor de vendas dele, caso ele tenha batido a meta

```python
from cores import *
vendedores_dic = {'Maria': 1200, 'José': 300, 'Antônio': 800, 'João': 1500, 'Francisco': 1900, 'Ana': 2750, 'Luiz': 400, 'Paulo': 20, 'Carlos': 23, 'Manoel': 70, 'Pedro': 90, 'Francisca': 80, 'Marcos': 1100, 'Raimundo': 999, 'Sebastião': 900, 'Antônia': 880, 'Marcelo': 870, 'Jorge': 50, 'Márcia': 1111, 'Geraldo': 120, 'Adriana': 300, 'Sandra': 450, 'Luis': 800}
meta = 1000
```

- Fazendo por for

```python
print(f"{Negrito}A meta de Vendas é:{Reset} {Verde}R${Reset}{Negrito}{meta}{Reset}\n")

for indice_cliente, (nome, venda) in enumerate(vendedores_dic.items()):
    if (venda>meta):
        bonus = venda + (venda*0.10)
        print(f"{Negrito}{indice_cliente+1}º Funcionário --> Nome: '{nome.capitalize()}' - Quanto de Bônus fez:{Reset} {Verde}R${Reset}{Negrito}{bonus-venda:.0f} - Totalizando:{Reset} {Verde}R${Reset}{Negrito}{bonus:.0f}{Reset}")
        print(f"{Verde}Atingiu a Meta✅{Reset}")
    else:
        venda = venda
        print(f"{Negrito}{indice_cliente+1}º Funcionário --> Nome: '{nome.capitalize()} Quanto fez:{Reset}{Verde}R${Reset}{Negrito}{venda} - Não conseguiu o Bônus{Reset}")
        print(f"{Vermelho}NÃO ATINGIU A META ❌{Reset}")
```

- Fazendo por List Comprehension

```python
print(f"{Negrito}A meta de Vendas é:{Reset} {Verde}R${Reset}{Negrito}{meta}{Reset}\n")

bateram_a_meta = list((nome,venda)for (nome,venda) in vendedores_dic.items() if venda>meta)

for indice_cliente, (nome, venda) in enumerate(bateram_a_meta):
    bonus = venda+(0.10*venda)
    print(f"{Negrito}{indice_cliente+1}º Funcionário --> Nome: '{nome.capitalize()}' - Quanto de Bônus fez:{Reset} {Verde}R${Reset}{Negrito}{bonus:.0f} - Totalizando:{Reset} {Verde}R${Reset}{Negrito}{bonus:.0f}{Reset}")
    print(f"{Verde}Atingiu a Meta✅{Reset}")
```
