# Métodos Format  
### (f-string ou `.format(parâmetros)`)

```python
from cores import (Azul, AmareloClaro, Magenta, Reset)

faturamento = 2000
custo = 500
lucro = faturamento - custo

print(
    f"{Magenta}O faturamento da loja foi:{Reset} {Verde}R${faturamento}{Reset} "
    f"{Magenta}O custo da loja foi:{Reset} {Verde}R${custo}{Reset}. "
    f"{Magenta}Assim, o lucro da loja foi:{Reset} {Verde}R${lucro}{Reset}"
)
```

## Método `f-string`
Forma moderna, mais legível e recomendada para formatação de textos em Python.
Forma utilizada anteriormente  
**Concatenação de textos**

```python
from cores import (Azul, AmareloClaro, Magenta, Verde, Reset)
f"""
{Azul}Método .format(parâmetros){Reset}
"""

print(
    "{magenta}O faturamento da loja foi:{reset} {verde}R${f}{reset} "
    "{magenta}O custo da loja foi:{reset} {verde}R${c}{reset}. "
    "{magenta}Assim, o lucro da loja foi:{reset} {verde}R${l}{reset}"
    .format(
        magenta=Magenta,
        verde=Verde,
        reset=Reset,
        f=faturamento,
        c=custo,
        l=lucro))
```

## Método `.format(parâmetros)`
Forma clássica de formatação utilizando placeholders `{}` e parâmetros nomeados ou posicionais.
