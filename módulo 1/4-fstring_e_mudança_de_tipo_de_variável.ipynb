## Exemplo com valores definidos no código

Aqui utilizamos valores já definidos para calcular o lucro.


```python
from cores import (Magenta, Reset)

faturamento = 1000
custo = 400

lucro = (faturamento - custo)

print(f"{Magenta}Com .format →{Reset} O faturamento foi de {faturamento} e o lucro de {lucro}".format(faturamento, lucro))


print(f"{Magenta}Com f-string →{Reset},O faturamento foi de {faturamento} e o lucro de {lucro}")
```

## Entrada de dados com `input()`

Agora o usuário informa os valores manualmente.


```python
from cores import (Verde, Azul, AmareloClaro, Magenta, Reset, Negrito)

faturamento = int(input(f"\n\n{Negrito}Insira o faturamento: R${Reset} "))
custo = int(input(f"\n\n{Negrito}Insira o custo: R${Reset} "))

lucro = (faturamento - custo)

print(f"\n{Magenta}Lucro calculado:{Reset} {Verde}R${lucro}{Reset}\n\n")
```
