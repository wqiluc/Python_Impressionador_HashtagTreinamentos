# Tipos de Variáveis em Python

Uma variável é um objeto — afinal, tudo em Python é um objeto.

## Principais tipos

- `int` → números inteiros  
- `str` → textos  
- `float` → números com casas decimais (ponto flutuante)  
- `bool` → valores lógicos (`True` ou `False`)

> ⚠️ **Observações**
>
> - Variáveis em Python não são declaradas explicitamente  
> - Evite usar nomes reservados da linguagem
