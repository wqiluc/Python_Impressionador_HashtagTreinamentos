## 💡 1. O que é um DataFrame?

É como se fosse uma **tabela**.

| Elemento do DataFrame 🔑 | Se comporta como 🔓 |
|------------------------|-------------------|
| Colunas | Chaves de um dicionário |
| Linhas | Itens de uma lista |

---

## 🔎 2. Selecionando dados no DataFrame

Considerando um DataFrame chamado `vendas_df`:

| Sintaxe 🔑 | O que faz 🔓 |
|---------|-----------|
| `vendas_df['coluna_x']` | Retorna os valores da coluna (um DataFrame de 1 coluna / `Series`) |
| `vendas_df[0]` | ❌ **NÃO** funciona assim para DataFrames |
| `vendas_df[:3]` | Pega as linhas até o índice 3 do DataFrame |
| `vendas_df[['coluna_x', 'coluna_y', 'coluna_z']]` | Cria um novo DataFrame só com as colunas `coluna_x`, `coluna_y` e `coluna_z` |
| `vendas_df['coluna_x'][0]` | Pega o item da 1ª linha da coluna `coluna_x` |

<br>

> ⚠️ **Cuidado:** diferente de uma lista, `vendas_df[0]` não busca a linha de índice 0 — colchetes com um único valor buscam **colunas**, não linhas.

### 🔧 4. Conhecendo os dados com `.info()`

O 1º passo de toda Análise de Dados é você entender o que existe na sua base de dados.

Usaremos o `.info()` para isso.

> 💡 **Dica:** `.info()` mostra o número de linhas, os nomes das colunas, quantos valores não-nulos cada uma tem e o tipo de dado (`dtype`) — ótimo para um primeiro diagnóstico da base.

### 🧵 5. Selecionando uma única coluna

Vamos criar então agora uma lista de Clientes.

```python
vendas_df['ID Cliente'][:21]
```

### 🧮 6. Selecionando múltiplas colunas

Vamos criar agora uma lista com os produtos e as quantidades de vendas dele, caso a gente queira analisar só os produtos (independente de data ou de cliente).

> 💡 **Dica:** para selecionar **mais de uma coluna**, passe uma lista de nomes dentro dos colchetes — `vendas_df[['coluna_x', 'coluna_y']]` — o resultado é um novo DataFrame.

```python
vendas_df[['ID Produto', 'Quantidade Vendida']][:21]
```
