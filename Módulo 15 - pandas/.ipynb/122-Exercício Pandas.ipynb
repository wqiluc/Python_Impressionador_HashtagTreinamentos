## 📥 2. Importando as Bases

3 arquivos, 3 jeitos de ler:

| Arquivo | Formato | Parâmetros usados | Motivo |
|---|---|---|---|
| `CadastroFuncionarios.csv` | csv | `sep=';', encoding='utf-8', decimal=','` | separador `;` e valores decimais com vírgula (ex: `574,04`) |
| `CadastroClientes.csv` | csv | `sep=';', encoding='utf-8'` | separador `;` |
| `BaseServiçosPrestados.xlsx` | xlsx | `read_excel` (padrão) | arquivo Excel, sem separador/encoding |

<br>

> 💡 **Dica:** o parâmetro `decimal=','` avisa o pandas que a vírgula é o separador decimal (padrão brasileiro) — sem ele, colunas como `Impostos` e `Beneficios` (que têm valores como `574,04`) viram texto (`object`) em vez de número.

```python
import pandas as pd
from cores import *

# importando as 3 bases do exercício

funcionarios_df = pd.read_csv(r'../spec/CadastroFuncionarios.csv', sep=';', encoding='utf-8', decimal=',')
clientes_df = pd.read_csv(r'../spec/CadastroClientes.csv', sep=';', encoding='utf-8')
servicos_df = pd.read_excel(r'../spec/BaseServiçosPrestados.xlsx')
```

```python
display(servicos_df.head())
```

## 💰 3. Valor Total da Folha Salarial

| Coluna 🔑 | O que é 🔓 |
|---|---|
| `Salario Base` | salário fixo do funcionário |
| `Beneficios` | valor de benefícios (plano de saúde, vale-refeição, etc) |
| `Impostos` | impostos pagos pela empresa sobre o funcionário |

Salário Total de cada funcionário = `Salario Base` + `Beneficios` + `Impostos`. Depois é só somar a coluna inteira.

```python
funcionarios_df['Salario Total'] = funcionarios_df['Salario Base'] + funcionarios_df['Beneficios'] + funcionarios_df['Impostos']

folha_salarial = funcionarios_df['Salario Total'].sum()

print(f"{CinzaClaro}Valor Total da Folha Salarial:{Reset} {VerdeClaro}R${folha_salarial:,.2f}{Reset}")
```

## 📈 4. Faturamento da Empresa

A base de serviços não tem o valor do contrato — quem tem é a base de clientes (`Valor Contrato Mensal`). Precisamos juntar as duas.

| Método 🔑 | O que faz 🔓 |
|---|---|
| `.merge(clientes_df, on='ID Cliente')` | traz o valor mensal do contrato de cada cliente para a base de serviços |
| `Valor Contrato Mensal * Tempo Total de Contrato (Meses)` | faturamento total daquele contrato específico |

```python
servicos_clientes_df = servicos_df.merge(clientes_df, on='ID Cliente')

servicos_clientes_df['Faturamento'] = servicos_clientes_df['Valor Contrato Mensal'] * servicos_clientes_df['Tempo Total de Contrato (Meses)']

faturamento_total = servicos_clientes_df['Faturamento'].sum()

print(f"{CinzaClaro}Faturamento Total da Empresa:{Reset} {VerdeClaro}R${faturamento_total:,.2f}{Reset}")
```

## 🤝 5. Percentual de Funcionários com Contrato Fechado

| Método 🔑 | O que faz 🔓 |
|---|---|
| `servicos_df['ID Funcionário'].unique()` | lista dos funcionários que aparecem na base de serviços, sem repetição |
| `len(...)` | quantidade de funcionários únicos |

% = Qtde_Funcionarios_Fecharam_Serviço / Qtde_Funcionários_Totais

```python
funcionarios_com_contrato = servicos_df['ID Funcionário'].unique()

qtde_funcionarios_com_contrato = len(funcionarios_com_contrato)
qtde_funcionarios_total = funcionarios_df.shape[0]

perc_funcionarios_com_contrato = qtde_funcionarios_com_contrato / qtde_funcionarios_total

print(f"{CinzaClaro}Funcionários com contrato fechado:{Reset} {VerdeClaro}{qtde_funcionarios_com_contrato}{Reset} {CinzaClaro}de{Reset} {VerdeClaro}{qtde_funcionarios_total}{Reset}")

print(f"{CinzaClaro}% de Funcionários que já Fecharam Contrato:{Reset} {VerdeClaro}{perc_funcionarios_com_contrato:.2%}{Reset}")
```

## 🏢 6. Contratos Fechados por Área

Cada linha de `servicos_df` é um contrato. Para saber a área de quem fechou, precisamos trazer a coluna `Area` (que só existe em `funcionarios_df`) via `merge`, e então agrupar.

| Método 🔑 | O que faz 🔓 |
|---|---|
| `.merge(funcionarios_df[['ID Funcionário', 'Area']], on='ID Funcionário')` | traz a área de cada funcionário para a base de serviços |
| `.groupby('Area')['Codigo do Servico'].count()` | conta quantos contratos (linhas) cada área fechou |

```python
servicos_area_df = servicos_df.merge(funcionarios_df[['ID Funcionário', 'Area']], on='ID Funcionário')

contratos_por_area = servicos_area_df.groupby('Area')['Codigo do Servico'].count()

display("Contratos por Área: ", contratos_por_area)
```

## 👥 7. Funcionários por Área

Aqui não precisa de `merge` — a própria base de funcionários já tem a coluna `Area`.

```python
funcionarios_por_area = funcionarios_df.groupby('Area')['ID Funcionário'].count()

display(funcionarios_por_area)
```

## 🎫 8. Ticket Médio Mensal dos Contratos

Usando a base já mesclada em [4. Faturamento da Empresa](#-4-faturamento-da-empresa) (`servicos_clientes_df`, 1 linha por contrato), o ticket médio mensal é a média da coluna `Valor Contrato Mensal`.

```python
ticket_medio_mensal = servicos_clientes_df['Valor Contrato Mensal'].mean()

print(f"{CinzaClaro}Ticket Médio Mensal dos Contratos:{Reset} {VerdeClaro}R${ticket_medio_mensal:,.2f}{Reset}")
```
