## ⚙️ 1. Preparando as bases de dados

- Preparando as bases de dados (o que fizemos nas últimas aulas)

````python
import pandas as pd

# importando os arquivos
vendas_df = pd.read_csv(r'../spec/Contoso - Vendas - 2017.csv', sep=';', encoding='latin1')
produtos_df = pd.read_csv(r'../spec/Contoso - Produtos Cadastro Completo.csv', sep=';', encoding='latin1')
lojas_df = pd.read_csv(r'../spec/Contoso - Lojas.csv', sep=';', encoding='latin1')
clientes_df = pd.read_csv(r'../spec/Contoso - Clientes.csv', sep=';', encoding='latin1')

# Ajustar nome das colunas
produtos_df.rename(columns={'ÿNome do Produto': 'Nome do Produto'}, inplace=True)
clientes_df.rename(columns={'ÿID Cliente': 'ID Cliente'}, inplace=True)
lojas_df.rename(columns={'ÿID Loja': 'ID Loja'}, inplace=True)

# limpando apenas as colunas que queremos
clientes_df = clientes_df[['ID Cliente', 'E-mail']]
produtos_df = produtos_df[['ID Produto', 'Nome do Produto']]
lojas_df = lojas_df[['ID Loja', 'Nome da Loja']]

# mesclando e renomeando os dataframes
vendas_df = vendas_df.merge(produtos_df, on='ID Produto')
vendas_df = vendas_df.merge(lojas_df, on='ID Loja')
vendas_df = vendas_df.merge(clientes_df, on='ID Cliente').rename(columns={'E-mail': 'E-mail do Cliente'})

#print 
display(vendas_df[0:21])
````

## 📊 2. Qual o % das vendas que foi devolvido?

Primeiro vamos aplicar uma função normalmente, sem nenhum filtro. Para isso vamos somar as quantidades nas colunas correspondentes. Lembrando, o % vai ser: `Total Devolvido / Total Vendido`.

| Método | O que faz |
|--------|-----------|
| `dataframe['coluna'].sum()` | Soma todos os valores da coluna |

````python
qtde_vendida = vendas_df['Quantidade Vendida'].sum()
qtde_devolvida = vendas_df['Quantidade Devolvida'].sum()

perc_devolucao = qtde_devolvida / qtde_vendida
print(f'{perc_devolucao:.2%}')
````

## 🏬 3. Filtrando por uma loja específica

Agora, se quisermos fazer a mesma análise apenas para 1 loja: queremos filtrar apenas os itens da Loja Contoso Europe Online e saber o % de devolução dessa loja.

Para isso, vamos precisar filtrar. A forma de filtrar um DataFrame é uma "simples" comparação:

```python
dataframe[dataframe['coluna'] == valor]
```

| Parte | Significado |
|-------|-------------|
| `dataframe['coluna'] == valor` | Cria uma condição: compara cada linha da coluna com o valor e retorna `True`/`False` |
| `dataframe[condição]` | Mantém apenas as linhas onde a condição é `True` |

> ⚠️ **Cuidado:** os nomes das lojas nessa base vêm com um espaço extra no final (ex.: `'Loja Contoso Europe Online '`), efeito da própria origem dos dados. Preste atenção na comparação exata.

````python
vendas_europe_df = vendas_df[vendas_df['Nome da Loja'] == 'Loja Contoso Europe Online ']

qtde_vendida_europe = vendas_europe_df['Quantidade Vendida'].sum()
qtde_devolvida_europe = vendas_europe_df['Quantidade Devolvida'].sum()

perc_devolucao_europe = (qtde_devolvida_europe / qtde_vendida_europe)

print(f'{perc_devolucao_europe:.2%}')
````

## 🪜 4. Entendendo o filtro em 2 passos

Vamos fazer em 2 passos para entender certinho o que está acontecendo por trás da comparação.

1. **Passo 1:** `vendas_df['Nome da Loja'] == 'Loja Contoso Europe Online '` sozinho já retorna uma coisa — uma `Series` de `True`/`False`, uma para cada linha do DataFrame (essa é a **condição**, ou **máscara booleana**).
2. **Passo 2:** usamos essa máscara dentro de `vendas_df[...]` para manter só as linhas marcadas como `True`.

````python
# Passo 1: criando a condição (máscara booleana)
condicao_europe = vendas_df['Nome da Loja'] == "Loja Contoso Europe Online"

display(condicao_europe[0:21])
````

````python
# Passo 2: usando a condição para filtrar o dataframe
vendas_europe_df = vendas_df[condicao_europe]

display(vendas_europe_df[0:21])
````

## 🧩 5. Desafio: filtrando com 2 condições

E se eu quisesse criar uma tabela apenas com as vendas da Loja Contoso Europe Online que não tiveram nenhuma devolução? Quero criar essa tabela e saber quantas vendas são.

Repare que nesse caso são 2 condições ao mesmo tempo — como fazemos isso?

| Operador | Significado |
|----------|-------------|
| `&` | **E** — as duas condições precisam ser verdadeiras |
| `\|` | **OU** — pelo menos uma das condições precisa ser verdadeira |

> ⚠️ **Cuidado:** ao combinar condições, cada uma precisa ficar entre parênteses: `dataframe[(condição1) & (condição2)]`. Sem os parênteses o Python não entende a prioridade das operações e dá erro.

````python
# tudo junto
vendas_europe_sem_devolucao_df = vendas_df[(vendas_df['Nome da Loja'] == 'Loja Contoso Europe Online ') & (vendas_df['Quantidade Devolvida'] == 0)]

display(vendas_europe_sem_devolucao_df[0:5]) #top 5

print(f'Total de vendas: {len(vendas_europe_sem_devolucao_df)}')
````

````python
# separado
condicao_loja = vendas_df['Nome da Loja'] == 'Loja Contoso Europe Online '
condicao_sem_devolucao = vendas_df['Quantidade Devolvida'] == 0

vendas_europe_sem_devolucao_df = vendas_df[condicao_loja & condicao_sem_devolucao]

display(vendas_europe_sem_devolucao_df)
print(f'Total de vendas: {len(vendas_europe_sem_devolucao_df)}')
````
