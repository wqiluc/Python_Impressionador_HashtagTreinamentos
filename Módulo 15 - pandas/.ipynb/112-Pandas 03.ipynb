## 💡 1. Por que usar o Pandas para ler CSV?

| Motivo 🔑 | Descrição 🔓 |
|--------|-----------|
| ⚡ Performance | Lê arquivos com centenas de milhares de linhas em segundos |
| 🧠 Praticidade | Uma única linha de código já transforma o `.csv` em uma tabela navegável |
| 🔧 Flexibilidade | Parâmetros como separador, encoding e tipos de coluna são configuráveis |

### 📌 Resumo

Quase sempre quando formos "ler" um arquivo csv, vamos usar o pandas. É prático e bem eficiente.

## ⚙️ 2. Estrutura do `read_csv`

Forma mais básica *(muitas vezes não usaremos a forma mais básica)*:

```python
dataframe = pd.read_csv(arquivo_com_extensao)
```

| Parâmetro 🔑 | Significado 🔓 |
|-----------|-------------|
| `arquivo_com_extensao` | Caminho do arquivo, incluindo a extensão `.csv` |
| `sep` | Separador usado entre as colunas (padrão é `,`, mas arquivos em pt-BR costumam usar `;`) |

> ⚠️ **Cuidado:** se o `.csv` usar `;` como separador e você não informar `sep=';'`, o pandas vai ler tudo como uma única coluna.

## 🔎 4. Explorando o DataFrame

| Método | O que retorna |
|--------|----------------|
| `.head()` | As primeiras linhas do DataFrame (5 por padrão) |
| `.shape` | Uma tupla `(linhas, colunas)` com o tamanho da tabela |
| `.dtypes` | O tipo de dado de cada coluna |

> 💡 **Dica:** rodar `.head()` logo depois de ler um arquivo é um ótimo hábito — confirma rapidamente se o `sep` e o encoding estavam corretos.
