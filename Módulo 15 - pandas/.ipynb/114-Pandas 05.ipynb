## 🎯 1. Objetivo

Vamos modificar os IDs para os nomes dos produtos, dos clientes e das lojas, para nossas análises ficarem mais intuitivas futuramente. Para isso, vamos criar um DataFrame com todos os detalhes.

- Usaremos o método `merge` para isso e, depois se quisermos, podemos pegar apenas as colunas que queremos do DataFrame final.

<br>

> ⚠️ **Cuidado:** os arquivos da Contoso usados aqui (`Clientes`, `Lojas`, `Produtos`) têm acentuação e às vezes precisam de um encoding diferente do padrão. Valores para testar: `encoding='latin1'`, `encoding='ISO-8859-1'`, `encoding='utf-8'` ou `encoding='cp1252'`.

---

## ⚙️ 2. Criando nossos DataFrames

- Vamos ler a base de Vendas e as bases de apoio (Clientes, Lojas e Produtos), cada uma com seu próprio encoding

````python
import pandas as pd
from cores import *

vendas_df = pd.read_csv("../spec/Contoso - Vendas - 2017.csv", sep=';')

display(vendas_df[0:21]) # Exibe as primeiras 21 linhas do DataFrame de vendas
````

````python
# às vezes precisaremos mudar o encoding — esses arquivos usam acentuação e vieram com encoding diferente do padrão utf-8

clientes_df = pd.read_csv("../spec/Contoso - Clientes.csv", sep=';', encoding='latin1')
lojas_df = pd.read_csv("../spec/Contoso - Lojas.csv", sep=';', encoding='latin1')
produtos_df = pd.read_csv("../spec/Contoso - Produtos Cadastro Completo.csv", sep=';', encoding='latin1')

display(clientes_df[0:21])
````

````python
display(lojas_df[0:21])
````

````python
display(produtos_df[0:21])
````

## 🧹 3. Tirando as coluna e infos inúteis ❌👎

Note que: `ID Cliente`, `ID Loja` e `Nome do Produto` vieram com um caractere estranho grudado no nome (efeito do encoding) — e o `clientes_df` ainda trouxe colunas vazias (`Unnamed: ...`). Vamos limpar cada DataFrame, renomeando a coluna de ID e mantendo só o que interessa para o merge.

| Sintaxe 🔑 | O que faz 🔓 |
|---------|-----------|
| `dataframe.drop(columns=['coluna1', 'coluna2'])` | Retira as colunas `coluna1` e `coluna2` |
| `dataframe[['coluna1', 'coluna2']]` | Mantém só as colunas informadas, descartando o resto |
| `dataframe.rename(columns={'nome_velho': 'nome_novo'})` | Renomeia colunas — útil quando o nome veio "sujo" |

<br>

> 💡 **Dica:** `dataframe[['col1', 'col2']]` costuma ser mais direto que `.drop()` quando você quer manter só 2 ou 3 colunas de uma tabela grande.

````python
clientes_df = clientes_df.rename(columns={'ÿID Cliente': 'ID Cliente'})
clientes_df = clientes_df.drop(columns=['Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9', 'Unnamed: 10'])

lojas_df = lojas_df.rename(columns={'ÿID Loja': 'ID Loja'})
lojas_df = lojas_df[['ID Loja', 'Nome da Loja']]

produtos_df = produtos_df.rename(columns={'ÿNome do Produto': 'Nome do Produto'})
produtos_df = produtos_df[['ID Produto', 'Nome do Produto']]

display(clientes_df[0:21])
````

````python
display(lojas_df[0:21])
````

````python
display(produtos_df[0:21])
````

## 🔗 4. Juntando os DataFrames com o método: "`merge`"

Agora vamos juntar os DataFrames para ter 1 único DataFrame com tudo "bonito":

```python
novo_dataframe = dataframe1.merge(dataframe2, on='coluna')
```

| Parte 🔑 | Significado 🔓 |
|-------|-------------|
| `dataframe1.merge(dataframe2, ...)` | Junta as colunas de `dataframe2` ao `dataframe1` |
| `on='coluna'` | A coluna em comum usada para casar as linhas dos dois DataFrames |

> ⚠️ **Cuidado:** o `merge` precisa das colunas com o **mesmo nome** para funcionar. Se não tiver, você precisa alterar o nome da coluna com o `.rename`:
>
> ```python
> dataframe.rename(columns={'coluna1': 'novo_coluna_1'})
> ```

````python
# juntando os dataframes

vendas_completo_df = vendas_df.merge(clientes_df, on='ID Cliente').merge(produtos_df, on='ID Produto').merge(lojas_df, on='ID Loja')

# exibindo o dataframe final

display(vendas_completo_df[0:21])
````

## ✏️ 5. Renomeando colunas com `rename`

Vamos renomear o `E-mail` para ficar claro que é do cliente — assim, se mais colunas de e-mail forem adicionadas no futuro (fornecedor, vendedor...), não haverá confusão sobre a quem cada uma pertence.

````python
# vamos renomear o e-mail para ficar claro que é do cliente
vendas_completo_df = vendas_completo_df.rename(columns={'E-mail': 'E-mail Cliente'})

display(vendas_completo_df[0:21])
````
